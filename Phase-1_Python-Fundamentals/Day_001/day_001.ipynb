## Part 1: Variables — Your Data Containers

Think of a variable like a **labeled jar** in your kitchen:

- The **label** on the jar is the variable name → `borehole_depth`
- The **contents** inside is the value → `15.5`
- You can **change** what's inside anytime → `borehole_depth = 20.0`

### Rules for naming variables:
- Must start with a letter or underscore: `depth`, `_temp`, `N_value` ✓
- Cannot start with a number: `1st_layer` ✗
- No spaces allowed: use underscores → `water_table_depth` ✓
- Case sensitive: `Depth` and `depth` are DIFFERENT variables

### Best practice:
Use **descriptive names**. Future-you (or your ML model code) will thank you.

| Bad Name | Good Name | Why |
|----------|-----------|-----|
| `x` | `borehole_depth` | You know exactly what it stores |
| `n` | `spt_n_value` | Clear it's an SPT blow count |
| `d` | `water_table_depth` | No guessing needed |

````python
# ─────────────────────────────────────────────────────
# PROJECT INFORMATION
# ─────────────────────────────────────────────────────
# These are all VARIABLES. Each one stores a piece of
# information about our borehole investigation.

project_name = "Highway Bridge Foundation Investigation"
project_id = "GE-2026-0042"
location = "Salt Lake City, Utah"
borehole_id = "BH-01"

# Let's print them to see what we stored
print("Project:", project_name)
print("ID:", project_id)
print("Location:", location)
print("Borehole:", borehole_id)
````

````python
# ─────────────────────────────────────────────────────
# BOREHOLE PARAMETERS
# ─────────────────────────────────────────────────────
# Notice: some values have decimals, some don't.
# Python automatically assigns the correct data type.

borehole_depth = 15.5          # Total depth drilled (meters)
water_table_depth = 3.2        # Where we hit water (meters)
number_of_samples = 8          # How many SPT samples collected
ground_elevation = 1320.75     # Elevation above sea level (meters)

print(f"Total Depth    : {borehole_depth} m")
print(f"Water Table    : {water_table_depth} m")
print(f"Samples Taken  : {number_of_samples}")
print(f"Ground Elevation: {ground_elevation} m")
````

````python
# ─────────────────────────────────────────────────────
# SPT DATA AT A SPECIFIC DEPTH
# ─────────────────────────────────────────────────────
# This represents one row of a borehole log sheet.

spt_n_value = 18               # SPT blow count (integer - always a whole number)
sample_depth = 3.0             # Depth of this sample (meters)
soil_type = "Silty Sand"       # What the soil looks like
is_saturated = False           # Is this sample below water table?

print(f"At depth {sample_depth}m:")
print(f"  Soil type  : {soil_type}")
print(f"  N-value    : {spt_n_value}")
print(f"  Saturated? : {is_saturated}")
````

### What just happened?

We created **12 variables** that store different kinds of data. Python figured out the type automatically:

```
project_name = "Highway Bridge..."    →  TEXT (string)
borehole_depth = 15.5                 →  DECIMAL NUMBER (float)
number_of_samples = 8                 →  WHOLE NUMBER (integer)
is_saturated = False                  →  YES/NO (boolean)
```

This is like filling out a borehole log form — each field has a specific kind of information.

## Part 2: Data Types — The Four Building Blocks

Python has **4 basic data types** that we use constantly:

| Type | Python Name | Example | Geotechnical Use |
|------|------------|---------|------------------|
| **Integer** | `int` | `18` | SPT N-value, number of layers |
| **Float** | `float` | `15.5` | Depth, unit weight, pressure |
| **String** | `str` | `"Silty Sand"` | Soil name, borehole ID |
| **Boolean** | `bool` | `True` / `False` | Is it saturated? Liquefiable? |

### Why does this matter?

When you build ML models later, the computer needs to know:
- **Numbers** (int/float) → can be used for math, plots, training
- **Text** (str) → needs to be *encoded* before ML can use it
- **Yes/No** (bool) → used for filtering, labeling, classification

Getting types wrong = bugs. Getting types right = smooth pipeline.

````python
# ─────────────────────────────────────────────────────
# CHECKING DATA TYPES with type()
# ─────────────────────────────────────────────────────
# The type() function tells you what kind of data a variable holds.
# Think of it like asking: "What kind of jar is this?"

print("Variable                Value                    Type")
print("─" * 65)
print(f"project_name            {project_name!r:25s} {type(project_name).__name__}")
print(f"borehole_depth          {str(borehole_depth):25s} {type(borehole_depth).__name__}")
print(f"number_of_samples       {str(number_of_samples):25s} {type(number_of_samples).__name__}")
print(f"spt_n_value             {str(spt_n_value):25s} {type(spt_n_value).__name__}")
print(f"soil_type               {soil_type!r:25s} {type(soil_type).__name__}")
print(f"is_saturated            {str(is_saturated):25s} {type(is_saturated).__name__}")
print(f"water_table_depth       {str(water_table_depth):25s} {type(water_table_depth).__name__}")
print(f"ground_elevation        {str(ground_elevation):25s} {type(ground_elevation).__name__}")
````

````python
# ─────────────────────────────────────────────────────
# WHY TYPES MATTER — Quick Demonstration
# ─────────────────────────────────────────────────────

# Math works with numbers (int + float)
total_pressure = 18.5 * 6.5   # unit_weight × depth
print(f"Overburden pressure: {total_pressure} kN/m² ✓")

# Joining works with strings
full_label = borehole_id + " at " + location
print(f"Label: {full_label} ✓")

# But mixing them WITHOUT casting causes errors!
# Uncomment the line below to see the error:
# broken = "Depth is " + 15.5    # ← TypeError!

# The fix: convert the number to string first
fixed = "Depth is " + str(15.5) + " meters"
print(f"{fixed} ✓")
````

### Key Insight

Python won't let you mix apples and oranges:

- `int + int` → works (addition)
- `float + float` → works (addition)
- `int + float` → works (Python auto-converts int to float)
- `str + str` → works (joins them together)
- `str + int` → **ERROR!** You must convert first.

This is where **type casting** saves us.

## Part 3: Type Casting — Converting Between Types

**Type casting** = telling Python to change data from one type to another.

| Function | What It Does | Example |
|----------|-------------|--------|
| `int()` | Convert to integer | `int("24")` → `24` |
| `float()` | Convert to float | `float("6.5")` → `6.5` |
| `str()` | Convert to string | `str(18)` → `"18"` |
| `bool()` | Convert to boolean | `bool(0)` → `False` |

### When do we need this in geotechnical work?

**All the time!** Here are real scenarios:

````python
# ─────────────────────────────────────────────────────
# SCENARIO 1: Field data arrives as TEXT
# ─────────────────────────────────────────────────────
# Your field engineer fills a digital form on a tablet.
# ALL form data comes as strings (text), even numbers!
# You MUST convert before doing math.

field_n_value = "24"          # Came from field form (string!)
field_depth = "6.5"           # Came from GPS logger (string!)
field_efficiency = "60"       # Hammer efficiency (string!)

print("Before casting:")
print(f"  N-value : '{field_n_value}' (type: {type(field_n_value).__name__})")
print(f"  Depth   : '{field_depth}' (type: {type(field_depth).__name__})")
print()

# Convert to proper types
N = int(field_n_value)          # String → Integer
depth = float(field_depth)      # String → Float
Ce = int(field_efficiency) / 100  # String → Int → Float (division makes it float)

print("After casting:")
print(f"  N-value : {N} (type: {type(N).__name__})")
print(f"  Depth   : {depth} (type: {type(depth).__name__})")
print(f"  Ce      : {Ce} (type: {type(Ce).__name__})")
````

````python
# ─────────────────────────────────────────────────────
# SCENARIO 2: Integer to Float (precision matters)
# ─────────────────────────────────────────────────────
# Field crew wrote depth as "7" instead of "7.0".
# We need float for precise calculations.

recorded_depth = 7                       # Integer (no decimal)
precise_depth = float(recorded_depth)    # Now it's 7.0

print(f"Recorded : {recorded_depth} (type: {type(recorded_depth).__name__})")
print(f"Precise  : {precise_depth} (type: {type(precise_depth).__name__})")
print()

# Why does this matter?
# In Python 3, 7 / 2 = 3.5 (correct)
# But in some older systems or ML libraries, 
# integer division can cause silent errors.
# Always use float for measurements!
````

````python
# ─────────────────────────────────────────────────────
# SCENARIO 3: Numbers to String (for reports & labels)
# ─────────────────────────────────────────────────────
# Building labels for borehole log reports.

layer_number = 3
depth_value = 4.5

# Method 1: Using str() + concatenation
label_v1 = "Layer-" + str(layer_number) + " at " + str(depth_value) + "m"
print(f"Method 1 (str + concat): {label_v1}")

# Method 2: Using f-strings (RECOMMENDED — cleaner!)
label_v2 = f"Layer-{layer_number} at {depth_value}m"
print(f"Method 2 (f-string)   : {label_v2}")

# Method 3: Using .format()
label_v3 = "Layer-{} at {}m".format(layer_number, depth_value)
print(f"Method 3 (.format)    : {label_v3}")

print()
print("→ Use f-strings (Method 2). They are the modern, clean way.")
````

````python
# ─────────────────────────────────────────────────────
# SCENARIO 4: Boolean Casting — Truthiness
# ─────────────────────────────────────────────────────
# In Python, some values are "truthy" and some are "falsy".
# This is VERY important for data filtering in ML.

# In geotechnical context:
# N=0 means weight-of-hammer (extremely soft soil, no resistance)
# N>0 means soil has some strength

n_zero = 0          # Weight-of-hammer
n_normal = 15       # Normal blow count
n_refusal = 50      # Refusal (very hard layer)

print(f"N = {n_zero}  → bool: {bool(n_zero)}   (No resistance detected!)")
print(f"N = {n_normal} → bool: {bool(n_normal)}  (Soil has strength)")
print(f"N = {n_refusal} → bool: {bool(n_refusal)}  (Refusal - very strong)")

print()
print("Falsy values in Python:")
print(f"  bool(0)     = {bool(0)}      (zero)")
print(f"  bool(0.0)   = {bool(0.0)}      (zero float)")
print(f"  bool('')    = {bool('')}      (empty string)")
print(f"  bool(None)  = {bool(None)}      (nothing)")
print()
print("Everything else is Truthy:")
print(f"  bool(1)     = {bool(1)}")
print(f"  bool(-5)    = {bool(-5)}")
print(f"  bool('hi')  = {bool('hi')}")
````

````python
# ─────────────────────────────────────────────────────
# SCENARIO 5: Float to Int (rounding/truncating)
# ─────────────────────────────────────────────────────
# Sometimes we need to round values.

calculated_n60 = 22.7    # Corrected N-value (float)

# int() TRUNCATES (cuts off decimal, does NOT round)
truncated = int(calculated_n60)
print(f"int(22.7) = {truncated}   ← Truncates (cuts off decimal)")

# round() ROUNDS properly
rounded = round(calculated_n60)
print(f"round(22.7) = {rounded}  ← Rounds (mathematically correct)")

# round() with decimal places
unit_weight = 18.5672
print(f"\nRaw unit weight: {unit_weight}")
print(f"round(18.5672, 1) = {round(unit_weight, 1)}   ← 1 decimal")
print(f"round(18.5672, 2) = {round(unit_weight, 2)}  ← 2 decimals")
````

### Type Casting Summary

```
String → Number:    int("24")     →  24
                    float("6.5")  →  6.5

Number → String:    str(18)       →  "18"
                    str(15.5)     →  "15.5"

Number → Boolean:   bool(0)       →  False
                    bool(15)      →  True

Float ↔ Integer:    int(22.7)     →  22    (truncates!)
                    float(7)      →  7.0
                    round(22.7)   →  23    (proper rounding)
```

## Part 4: Geotechnical Application
### SPT Borehole Log — Complete Workflow

Now let's combine everything we learned into a **real geotechnical workflow**:

1. Receive raw field data (strings from digital forms)
2. Cast to proper types
3. Perform SPT corrections
4. Estimate soil properties
5. Generate a formatted report

This is the exact pipeline that feeds into ML models later:

```
Raw Field Data (strings) → Type Casting → Clean Numbers → Calculations → ML Model Input
```

````python
# ══════════════════════════════════════════════════════
# COMPLETE SPT ANALYSIS — Putting It All Together
# ══════════════════════════════════════════════════════

# ── STEP 1: Raw field data (all strings, as received) ──
raw_borehole = "BH-01"
raw_depth = "6.5"
raw_n_value = "22"
raw_efficiency = "60"
raw_soil = "Silty Sand (SM)"
raw_water_table = "3.2"
raw_unit_weight = "18.5"

print("STEP 1: Raw field data received")
print(f"  All values are strings: type(raw_n_value) = {type(raw_n_value).__name__}")
print()
````

````python
# ── STEP 2: Type Casting ──
bh_id = raw_borehole                     # Already a string (keep it)
depth = float(raw_depth)                  # String → Float
N = int(raw_n_value)                      # String → Integer
Ce = int(raw_efficiency) / 100            # String → Int → Float
soil = raw_soil                           # Already a string (keep it)
wt_depth = float(raw_water_table)         # String → Float
gamma = float(raw_unit_weight)            # String → Float (unit weight)

print("STEP 2: Type casting complete")
print(f"  depth = {depth} ({type(depth).__name__})")
print(f"  N     = {N} ({type(N).__name__})")
print(f"  Ce    = {Ce} ({type(Ce).__name__})")
print(f"  gamma = {gamma} ({type(gamma).__name__})")
print()
````

````python
# ── STEP 3: SPT N-value Correction (N60) ──
# The raw SPT N-value must be corrected for hammer efficiency.
# Formula: N60 = N × (Ce / 0.60)
#
# Ce = Hammer energy ratio (typically 0.60 for safety hammer)
# This correction standardizes N-values across different equipment.

N60 = N * (Ce / 0.60)

print("STEP 3: SPT N-value Correction")
print(f"  N (field)  = {N}")
print(f"  Ce         = {Ce}")
print(f"  N60        = {N} x ({Ce}/0.60) = {N60:.1f}")
print()
````

````python
# ── STEP 4: Estimate Soil Properties ──

# 4a: Overburden Pressure
# σv = γ × depth
sigma_v = gamma * depth
print("STEP 4a: Overburden Pressure")
print(f"  σv = {gamma} × {depth} = {sigma_v:.1f} kN/m²")
print()

# 4b: Shear Wave Velocity (Imai & Tonouchi, 1982)
# Vs = 97 × N60^0.314
Vs = 97 * (N60 ** 0.314)
print("STEP 4b: Shear Wave Velocity (Vs)")
print(f"  Vs = 97 × {N60:.1f}^0.314 = {Vs:.1f} m/s")
print()

# 4c: Relative Density (Meyerhof, 1957 — for sands)
# Dr(%) ≈ 21 × √(N60)    (simplified)
Dr = 21 * (N60 ** 0.5)
print("STEP 4c: Relative Density (Dr)")
print(f"  Dr = 21 × √{N60:.1f} = {Dr:.1f}%")
print()

# 4d: Soil Consistency Classification
if N60 < 4:
    consistency = "Very Loose"
elif N60 < 10:
    consistency = "Loose"
elif N60 < 30:
    consistency = "Medium Dense"
elif N60 < 50:
    consistency = "Dense"
else:
    consistency = "Very Dense"

print("STEP 4d: Soil Consistency")
print(f"  N60 = {N60:.1f} → {consistency}")
print()

# 4e: Saturation Check
is_saturated = depth > wt_depth    # Boolean result!
print("STEP 4e: Saturation Check")
print(f"  Sample depth ({depth}m) > Water table ({wt_depth}m)? → {is_saturated}")
````

````python
# ── STEP 5: Generate Formatted Report ──
# Here we use str() and f-strings to build a clean output.

print("\n" + "=" * 55)
print("  SPT ANALYSIS REPORT")
print("=" * 55)
print(f"  Borehole     : {bh_id}")
print(f"  Depth        : {depth} m")
print(f"  Soil Type    : {soil}")
print(f"  Water Table  : {wt_depth} m")
print("─" * 55)
print(f"  SPT N (field): {N}")
print(f"  N60          : {N60:.1f}")
print(f"  σv           : {sigma_v:.1f} kN/m²")
print(f"  Vs           : {Vs:.1f} m/s")
print(f"  Dr           : {Dr:.1f}%")
print(f"  Consistency  : {consistency}")
print(f"  Saturated    : {'Yes' if is_saturated else 'No'}")
print("─" * 55)

if is_saturated and N60 < 20:
    print("  WARNING: Potentially liquefiable layer!")
    print("  → Detailed liquefaction analysis recommended.")
else:
    print("  STATUS: No immediate liquefaction concern.")

print("=" * 55)
````

## Key Takeaways

1. **Variables** are named containers for data. Use descriptive names like `spt_n_value`, not `x`.

2. **Four basic types**: `int` (whole numbers), `float` (decimals), `str` (text), `bool` (True/False). Each has its purpose in geotechnical data.

3. **Type casting** with `int()`, `float()`, `str()`, `bool()` converts between types. This is critical because field data often arrives as strings.

4. **f-strings** (`f"N = {value}"`) are the cleanest way to mix text and numbers.

5. **The geotechnical pipeline**: Raw strings from field → Cast to numbers → Calculate → Generate report. This same pipeline feeds ML models in later phases.
