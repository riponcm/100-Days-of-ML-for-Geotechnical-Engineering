## Our Geotechnical Problem

Borehole data is fundamentally a **collection** of values:

```
Depths:      [1.5, 3.0, 4.5, 6.0, ...]       ← data that grows (LIST)
Sieve sizes: (75.0, 19.0, 4.75, 2.0, ...)     ← fixed standard (TUPLE)
Soil types:  {"Sand", "Clay", "Gravel"}        ← unique types found (SET)
```

Each collection type has a **specific purpose**:

| Collection | Ordered? | Changeable? | Duplicates? | Best For |
|-----------|----------|-------------|-------------|----------|
| **List** `[]` | Yes | Yes (mutable) | Yes | Data that grows or changes |
| **Tuple** `()` | Yes | No (immutable) | Yes | Constants, coordinates |
| **Set** `{}` | No | Yes | No (unique only) | Distinct values, comparisons |

Today we master all three — these are the **building blocks of every dataset**.

## Part 1: Creating Lists — Your First Data Container

A **list** is an ordered collection of items inside square brackets `[]`. Think of it as a **column in a spreadsheet** — a sequence of values that you can access by position.

```python
# Each item has a position (index) starting from 0
n_values = [5, 8, 12, 18, 28, 35, 50, 60]
#           ↑  ↑   ↑   ↑   ↑   ↑   ↑   ↑
# Index:    0  1   2   3   4   5   6   7
```

Lists can hold **any type** of data: integers, floats, strings, booleans, even other lists.

````python
# ─────────────────────────────────────────────────────
# CREATING LISTS — Borehole data as collections
# ─────────────────────────────────────────────────────

# SPT data — each list is one "column" of our borehole log
depths = [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]       # floats
n_values = [5, 8, 12, 18, 28, 35, 50, 60]                    # integers
soil_types = ["Fill", "Silty Sand", "Sand", "Clayey Sand",   # strings
              "Sand", "Gravel", "Weathered Rock", "Bedrock"]
is_saturated = [False, False, True, True, True, True, True, True]  # booleans

print(f"Depths:     {depths}")
print(f"N-values:   {n_values}")
print(f"Soils:      {soil_types}")
print(f"Saturated:  {is_saturated}")
print()
print(f"Number of layers: {len(depths)}")
print(f"Type: {type(depths)}")
````

````python
# ─────────────────────────────────────────────────────
# SPECIAL WAYS TO CREATE LISTS
# ─────────────────────────────────────────────────────

# Empty list — start collecting before drilling begins
field_readings = []
print(f"Empty list: {field_readings}, length = {len(field_readings)}")

# List from range()
sample_numbers = list(range(1, 9))     # [1, 2, 3, ..., 8]
print(f"From range: {sample_numbers}")

# List of repeated values
zeros = [0] * 8                         # [0, 0, 0, 0, 0, 0, 0, 0]
print(f"Repeated zeros: {zeros}")

# List from a string (split)
csv_text = "5,8,12,18,28,35,50,60"
from_csv = csv_text.split(",")          # List of strings!
print(f"From CSV split: {from_csv}")
print(f"  Note: These are strings, not numbers!")

# Convert to integers
from_csv_int = [int(x) for x in from_csv]
print(f"  Converted: {from_csv_int}")
````

## Part 2: Indexing & Slicing — Accessing Specific Layers

Every item in a list has a **position number** (index). This works exactly like string indexing from Day 004.

```
List:     [5,   8,   12,  18,  28,  35,  50,  60]
Index:     0    1    2    3    4    5    6    7
Negative: -8   -7   -6   -5   -4   -3   -2   -1
```

**Indexing:** Access ONE item → `list[index]` 
**Slicing:** Access a RANGE → `list[start:stop:step]` (stop not included)

````python
# ─────────────────────────────────────────────────────
# INDEXING — Access individual layers
# ─────────────────────────────────────────────────────

n_values = [5, 8, 12, 18, 28, 35, 50, 60]

print("INDEXING (single item):")
print(f"  n_values[0]  = {n_values[0]}   ← first item (surface layer)")
print(f"  n_values[3]  = {n_values[3]}   ← 4th item (index 3)")
print(f"  n_values[7]  = {n_values[7]}   ← 8th item (last)")
print(f"  n_values[-1] = {n_values[-1]}   ← last item (deepest layer)")
print(f"  n_values[-2] = {n_values[-2]}   ← second to last")
print()

# Access corresponding data across parallel lists
depths = [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]
soil_types = ["Fill", "Silty Sand", "Sand", "Clayey Sand",
              "Sand", "Gravel", "Weathered Rock", "Bedrock"]

i = 3  # 4th layer
print(f"Layer {i+1} details (index {i}):")
print(f"  Depth: {depths[i]}m")
print(f"  N60:   {n_values[i]}")
print(f"  Soil:  {soil_types[i]}")
````

````python
# ─────────────────────────────────────────────────────
# SLICING — Access ranges of data
# ─────────────────────────────────────────────────────
# Syntax: list[start:stop]    (stop NOT included)
# Syntax: list[start:stop:step]

n_values = [5, 8, 12, 18, 28, 35, 50, 60]
print(f"Full list: {n_values}")
print()

print("Slicing examples:")
print(f"  [:3]    = {n_values[:3]}        ← first 3 layers (shallow zone)")
print(f"  [-3:]   = {n_values[-3:]}     ← last 3 layers (deep zone)")
print(f"  [2:6]   = {n_values[2:6]}   ← layers 3-6 (middle zone)")
print(f"  [::2]   = {n_values[::2]}    ← every other layer")
print(f"  [1::2]  = {n_values[1::2]}    ← every other, start from 2nd")
print(f"  [::-1]  = {n_values[::-1]}  ← reversed (bottom to top)")
print()

print("Practical uses:")
print(f"  Shallow zone (top 3):  avg N = {sum(n_values[:3])/3:.1f}")
print(f"  Deep zone (bottom 3):  avg N = {sum(n_values[-3:])/3:.1f}")
print(f"  N increases with depth — as expected for normally consolidated soil!")
````

### Indexing & Slicing Cheat Sheet

| Syntax | Meaning | With `[5, 8, 12, 18, 28]` |
|--------|---------|---------------------------|
| `l[0]` | First item | `5` |
| `l[-1]` | Last item | `28` |
| `l[:3]` | First 3 | `[5, 8, 12]` |
| `l[2:]` | From index 2 to end | `[12, 18, 28]` |
| `l[1:4]` | Index 1, 2, 3 | `[8, 12, 18]` |
| `l[-2:]` | Last 2 | `[18, 28]` |
| `l[::2]` | Every 2nd item | `[5, 12, 28]` |
| `l[::-1]` | Reversed | `[28, 18, 12, 8, 5]` |

## Part 3: Modifying Lists — Lists are MUTABLE

Unlike strings (Day 004) and tuples (later today), lists **can be changed after creation**. This is called being **mutable**.

You can:
- Change individual items: `list[0] = new_value`
- Change a range: `list[1:3] = [new1, new2]`
- Add items: `list.append(item)`
- Remove items: `list.pop(index)`

This is why lists are perfect for **field data** — you can update readings as corrections come in.

````python
# ─────────────────────────────────────────────────────
# MODIFYING LISTS — Change items after creation
# ─────────────────────────────────────────────────────

n_copy = [5, 8, 12, 18, 28, 35, 50, 60]
print(f"Original: {n_copy}")

# Change a single value (correction from field)
n_copy[0] = 6  # Corrected first reading
print(f"After n_copy[0] = 6:  {n_copy}")

# Change a range
n_copy[1:3] = [10, 14]  # Updated layers 2-3
print(f"After n_copy[1:3] = [10, 14]: {n_copy}")
print()

# Compare to strings (Day 004):
# my_string = "Hello"
# my_string[0] = "J"  ← TypeError! Strings are IMMUTABLE.
#
# my_list = ["H", "e", "l", "l", "o"]
# my_list[0] = "J"    ← Works! Lists are MUTABLE.

print("Lists are MUTABLE → can change items in place.")
print("Strings are IMMUTABLE → must create a new string.")
````

## Part 4: List Methods — Building Data as Drilling Continues

List methods modify the list **in place**. Here are the essential ones:

| Method | What It Does | Example |
|--------|-------------|--------|
| `.append(x)` | Add ONE item to end | `data.append(22)` |
| `.extend([x,y])` | Add MULTIPLE items | `data.extend([35, 50])` |
| `.insert(i, x)` | Insert at position i | `data.insert(0, 2)` |
| `.pop(i)` | Remove & return item at i | `data.pop(0)` |
| `.pop()` | Remove & return last item | `data.pop()` |
| `.remove(x)` | Remove first occurrence of x | `data.remove(8)` |
| `.sort()` | Sort in place | `data.sort()` |
| `.reverse()` | Reverse in place | `data.reverse()` |
| `.count(x)` | Count occurrences | `data.count(8)` |
| `.index(x)` | Find position of x | `data.index(22)` |

````python
# ─────────────────────────────────────────────────────
# LIST METHODS — Simulating field data collection
# ─────────────────────────────────────────────────────
# Imagine you're at the drill site. N-values arrive one by one.

field_n = []  # Empty — drilling hasn't started
print(f"Start:             {field_n}")

# Morning: Layer results arrive one at a time
field_n.append(5)     # append — add to the END
field_n.append(8)
field_n.append(12)
print(f"After 3 appends:   {field_n}")

# Afternoon: Bulk results from lab
field_n.extend([18, 28, 35])  # extend — add MULTIPLE items
print(f"After extend:      {field_n}")

# Oops! Forgot to log the surface layer
field_n.insert(0, 2)  # insert — add at specific position
print(f"After insert(0,2): {field_n}")

# That surface reading was wrong — remove it
removed = field_n.pop(0)  # pop — remove and return
print(f"Popped {removed}:          {field_n}")

# Remove last item (end of day, one reading was garbage)
last = field_n.pop()  # pop() with no index = remove last
print(f"Popped last ({last}):   {field_n}")
````

````python
# ─────────────────────────────────────────────────────
# SORTING and USEFUL OPERATIONS
# ─────────────────────────────────────────────────────

n_values = [5, 8, 12, 18, 28, 35, 50, 60]

# Built-in functions (DON'T modify the original)
print("Built-in functions:")
print(f"  len():    {len(n_values)} layers")
print(f"  min():    {min(n_values)}  (weakest layer)")
print(f"  max():    {max(n_values)}  (strongest layer)")
print(f"  sum():    {sum(n_values)}")
print(f"  average:  {sum(n_values)/len(n_values):.1f}")
print()

# sorted() vs .sort()
readings = [28, 5, 18, 12, 35, 8]
print(f"Original:    {readings}")
print(f"sorted():    {sorted(readings)}   ← returns NEW list")
print(f"Original:    {readings}            ← unchanged!")
print()

readings.sort()  # Modifies IN PLACE
print(f".sort():     {readings}          ← changed!")
readings.sort(reverse=True)
print(f"Descending:  {readings}")
print()

# count and index
soils = ["Sand", "Clay", "Sand", "Gravel", "Sand"]
print(f"Soil list:   {soils}")
print(f"  count('Sand'):  {soils.count('Sand')} times")
print(f"  index('Clay'):  position {soils.index('Clay')}")
````

## Part 5: List Comprehension — The Pythonic Superpower

List comprehension creates a new list by **transforming every item in one line**. It replaces a 4-line `for` loop with 1 line.

```python
# Traditional loop (4 lines):
result = []
for n in n_values:
    vs = 97 * (n ** 0.314)
    result.append(vs)

# List comprehension (1 line!):
result = [97 * (n ** 0.314) for n in n_values]
```

### Syntax Breakdown

```python
[expression  FOR  variable  IN  iterable]
 ↑ what to     ↑ each       ↑ source
   compute       item         data
```

This is one of **the most important Python features** for data science. Master it today — you'll use it every single day.

````python
# ─────────────────────────────────────────────────────
# LIST COMPREHENSION — Transform all layers at once
# ─────────────────────────────────────────────────────

n_values = [5, 8, 12, 18, 28, 35, 50, 60]
depths = [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]

# Shear wave velocity: Vs = 97 × N60^0.314
vs_values = [97 * (n ** 0.314) for n in n_values]
print(f"Vs (m/s): {[round(v,1) for v in vs_values]}")

# Overburden pressure: σv = γ × z
sigma_v = [18.5 * d for d in depths]
print(f"σv (kPa): {[round(s,1) for s in sigma_v]}")

# Relative density: Dr = 21 × √N60 (%)
Dr = [21 * (n ** 0.5) for n in n_values]
print(f"Dr (%):   {[round(d,1) for d in Dr]}")

# Quick classification labels
labels = ["Weak" if n < 15 else "Strong" for n in n_values]
print(f"Labels:   {labels}")

# Convert numbers to strings (for CSV output)
n_strings = [str(n) for n in n_values]
print(f"Strings:  {n_strings}")
print()
print("Each one-liner replaces a 4-line for loop!")
````

## Part 6: Filtered Comprehension — With Conditions

Add an `if` at the end to **filter** which items to include:

```python
[expression  FOR  variable  IN  iterable  IF  condition]
```

This is like SQL's `SELECT ... WHERE ...` or Excel's `FILTER` — pick only the items that meet a condition.

````python
# ─────────────────────────────────────────────────────
# FILTERED COMPREHENSION — Select specific layers
# ─────────────────────────────────────────────────────

n_values = [5, 8, 12, 18, 28, 35, 50, 60]
depths =   [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]
soils =    ["Fill", "Silty Sand", "Sand", "Clayey Sand",
            "Sand", "Gravel", "Weathered Rock", "Bedrock"]
wt = 3.2

# Only strong layers (N >= 20)
strong = [n for n in n_values if n >= 20]
print(f"Strong (N>=20):     {strong}")

# Only weak layers (N < 15)
weak = [n for n in n_values if n < 15]
print(f"Weak (N<15):        {weak}")

# Saturated depths (below water table)
sat_depths = [d for d in depths if d > wt]
print(f"Saturated depths:   {sat_depths}")

# N-values of sandy layers only (using zip)
sandy_n = [n for n, s in zip(n_values, soils) if "Sand" in s]
print(f"Sandy layer N:      {sandy_n}")

# Complex: risky layers (saturated + weak + sandy)
risky = [(d, n, s) for d, n, s in zip(depths, n_values, soils)
         if d > wt and n < 20 and "Sand" in s]
print(f"Liquefaction risk:  {risky}")

# Count with comprehension
weak_count = len([n for n in n_values if n < 15])
print(f"Weak layer count:   {weak_count}")
````

## Part 7: Tuples — Data That Should Never Change

A **tuple** is like a list but **immutable** — once created, it **cannot** be modified. Created with parentheses `()` instead of square brackets `[]`.

```python
# List — can change
n_list = [5, 8, 12]
n_list[0] = 6        # Works!

# Tuple — cannot change
coords = (40.76, -111.89)
coords[0] = 41.0     # TypeError! Tuples are immutable.
```

### Why use tuples?

They **protect data from accidental changes**. Borehole coordinates, sieve sizes, and engineering constants should never be modified after they're recorded.

````python
# ─────────────────────────────────────────────────────
# TUPLES — Immutable geotechnical constants
# ─────────────────────────────────────────────────────

# Borehole coordinates — should NEVER change after surveying
bh01_location = (40.7608, -111.8910, 1320.5)  # (lat, lon, elevation)

print(f"BH-01 coordinates: {bh01_location}")
print(f"Type: {type(bh01_location)}")
print()

# Indexing works the same as lists
print(f"Latitude:  {bh01_location[0]}")
print(f"Longitude: {bh01_location[1]}")
print(f"Elevation: {bh01_location[2]}m")
print()

# UNPACKING — assign each element to a variable
lat, lon, elev = bh01_location
print(f"Unpacked: lat={lat}, lon={lon}, elev={elev}m")
print()

# Standard sieve sizes (ASTM) — fixed engineering standard
sieve_sizes_mm = (75.0, 19.0, 4.75, 2.0, 0.425, 0.075)
print(f"Sieve sizes: {sieve_sizes_mm}")
print(f"Finest sieve: {sieve_sizes_mm[-1]}mm (No. 200)")
````

````python
# ─────────────────────────────────────────────────────
# TUPLE OF TUPLES — Read-only data table
# ─────────────────────────────────────────────────────

# Multiple borehole coordinates as a tuple of tuples
borehole_coords = (
    ("BH-01", 40.7608, -111.8910),
    ("BH-02", 40.7615, -111.8895),
    ("BH-03", 40.7602, -111.8920),
)

print("Borehole Coordinates (read-only):")
for name, lat, lon in borehole_coords:
    print(f"  {name}: ({lat}, {lon})")

print()

# This would FAIL — and that's a GOOD thing:
# bh01_location[0] = 41.0       # TypeError!
# sieve_sizes_mm.append(0.01)   # AttributeError!
# borehole_coords[0] = ("BH-04", 40.77, -111.90)  # TypeError!

print("Tuples PROTECT data:")
print("  Can't change coordinates → no accidental GPS errors")
print("  Can't change sieve sizes → ensures standard compliance")
print("  Can't modify survey data → maintains field record integrity")
````

### List vs Tuple — Decision Guide

| Question | Answer → Use |
|----------|-------------|
| Will data grow? (new samples) | **List** |
| Will items be updated? (corrections) | **List** |
| Should data be protected? (coordinates) | **Tuple** |
| Is it a fixed standard? (sieve sizes) | **Tuple** |
| Returning multiple values from function? | **Tuple** |

**Rule of thumb:** If it came from the field and might change → List. If it's a constant or standard → Tuple.

## Part 8: Sets — Unique Values & Comparisons

A **set** is an **unordered** collection of **unique** items. Duplicates are automatically removed.

```python
soils = {"Sand", "Clay", "Sand", "Gravel", "Sand"}
# Result: {"Sand", "Clay", "Gravel"}  — only 3 unique types
```

### Key Properties

| Property | List | Tuple | Set |
|----------|------|-------|-----|
| Ordered | Yes | Yes | **No** |
| Duplicates | Yes | Yes | **No** |
| Mutable | Yes | No | Yes |
| Indexing | `l[0]` | `t[0]` | **Not possible** |
| Use for | Sequential data | Constants | Unique values |

````python
# ─────────────────────────────────────────────────────
# SETS — Finding unique values
# ─────────────────────────────────────────────────────

# What unique soil types did we encounter in the borehole?
soil_log = ["Fill", "Sand", "Clay", "Sand", "Gravel",
            "Sand", "Clay", "Sand", "Gravel", "Bedrock"]

print(f"Full log ({len(soil_log)} entries): {soil_log}")

# Convert to set — duplicates removed!
unique_soils = set(soil_log)
print(f"Unique ({len(unique_soils)} types):   {unique_soils}")
print()

# Creating sets directly
test = {1, 2, 3, 2, 1, 3, 4}
print(f"{{1,2,3,2,1,3,4}} → {test}  (duplicates gone!)")
print()

# Note: Sets are UNORDERED — no indexing!
# unique_soils[0]  ← TypeError! Sets don't support indexing.
# If you need order, convert back to sorted list:
sorted_soils = sorted(unique_soils)
print(f"As sorted list: {sorted_soils}")
````

````python
# ─────────────────────────────────────────────────────
# SET OPERATIONS — Compare boreholes
# ─────────────────────────────────────────────────────
# Set operations answer questions like:
# "What soils are common across all boreholes?"
# "What's unique to this borehole?"

bh01_soils = {"Fill", "Sand", "Silty Sand", "Clay", "Gravel", "Bedrock"}
bh02_soils = {"Fill", "Sand", "Clay", "Silt", "Weathered Rock"}
bh03_soils = {"Sand", "Gravel", "Cobbles", "Bedrock"}

print(f"BH-01: {bh01_soils}")
print(f"BH-02: {bh02_soils}")
print(f"BH-03: {bh03_soils}")
print()

# UNION | — All soils found at the site (everything combined)
all_soils = bh01_soils | bh02_soils | bh03_soils
print(f"UNION (all types at site):")
print(f"  {all_soils}")
print(f"  Total: {len(all_soils)} unique types")
print()

# INTERSECTION & — Soils found in ALL boreholes
common = bh01_soils & bh02_soils & bh03_soils
print(f"INTERSECTION (common to all): {common}")
print()

# DIFFERENCE - — Soils in BH-01 but NOT in BH-02
only_bh01 = bh01_soils - bh02_soils
print(f"DIFFERENCE (only in BH-01): {only_bh01}")
print()

# SYMMETRIC DIFFERENCE ^ — In one OR the other, but NOT both
unique_either = bh01_soils ^ bh02_soils
print(f"SYMMETRIC DIFF (exclusive): {unique_either}")
````

````python
# ─────────────────────────────────────────────────────
# PRACTICAL SET USES
# ─────────────────────────────────────────────────────

# Use Case 1: Remove duplicate borehole IDs
field_ids = ["BH-01", "BH-02", "BH-01", "BH-03", "BH-02", "BH-01"]
unique_ids = sorted(set(field_ids))
print(f"All IDs:    {field_ids}")
print(f"Unique IDs: {unique_ids}")
print()

# Use Case 2: Validate entries against allowed values
allowed = {"Sand", "Silt", "Clay", "Gravel", "Fill", "Bedrock"}
entered = {"Sand", "Mud", "Clay", "Muck"}
valid = entered & allowed       # Intersection
invalid = entered - allowed     # Difference
print(f"Valid entries:   {valid}")
print(f"Invalid entries: {invalid} ← flag these!")
print()

# Use Case 3: Track testing progress
planned = {1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0}
tested = {1.5, 3.0, 4.5, 6.0, 7.5}
remaining = planned - tested
print(f"Planned:   {sorted(planned)}")
print(f"Tested:    {sorted(tested)}")
print(f"Remaining: {sorted(remaining)} ← still need these depths")

progress = len(tested) / len(planned) * 100
print(f"Progress: {progress:.0f}%")
````

### Set Operations Cheat Sheet

| Operation | Symbol | Method | Meaning |
|-----------|--------|--------|--------|
| Union | `A \| B` | `A.union(B)` | Everything in A or B or both |
| Intersection | `A & B` | `A.intersection(B)` | Only items in BOTH A and B |
| Difference | `A - B` | `A.difference(B)` | Items in A but NOT in B |
| Symmetric Diff | `A ^ B` | `A.symmetric_difference(B)` | Items in one but not both |
| Membership | `x in A` | — | Is x in the set? (very fast!) |

## Part 9: Nested Lists — 2D Data (Like a Spreadsheet)

## Part 10: Complete Application — Data Pipeline

````python
# ══════════════════════════════════════════════════════
# COMPLETE: BOREHOLE DATA PIPELINE
# Using Lists, Tuples, Sets, and Comprehensions
# ══════════════════════════════════════════════════════

# LISTS — mutable data from field
depths = [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]
n_raw =  [5, 8, 12, 18, 28, 35, 50, 60]
soils =  ["Fill", "Silty Sand", "Sand", "Clayey Sand",
          "Sand", "Gravel", "Weathered Rock", "Bedrock"]

# TUPLE — fixed site information
site_info = ("BH-01", 40.7608, -111.8910, "Salt Lake City, Utah")
wt = 3.2

# COMPREHENSIONS — compute derived values
n60 = [n * (60/60) for n in n_raw]
vs = [97 * (n ** 0.314) for n in n60]
sigma_v = [18.5 * d for d in depths]
saturated = [d > wt for d in depths]

def classify(n):
    if n < 4: return "V.Loose"
    elif n < 10: return "Loose"
    elif n < 30: return "Med Dense"
    elif n < 50: return "Dense"
    else: return "V.Dense"

classes = [classify(n) for n in n60]

# SET — unique soil types
unique_soils = set(soils)

# FILTERED COMPREHENSION — risky layers
risky = [(d, n, s) for d, n, s in zip(depths, n60, soils)
         if d > wt and n < 20 and "Sand" in s]

# Unpack site info tuple
bh_name, lat, lon, city = site_info

# Report
print("=" * 75)
print(f"  {bh_name} | {city} ({lat}, {lon}) | WT: {wt}m")
print(f"  Unique soils: {sorted(unique_soils)}")
print("=" * 75)
print(f"  {'#':>2} {'Dep':>5} {'N60':>4} {'Soil':16s} {'Class':9s} {'Vs':>6} {'σv':>6} Sat")
print("─" * 75)

for i, (d, n, s, c, v, sv, st) in enumerate(
    zip(depths, n60, soils, classes, vs, sigma_v, saturated), start=1):
    flag = "Y" if st else "N"
    print(f"  {i:2d} {d:4.1f}m {n:4.0f} {s:16s} {c:9s} {v:5.1f} {sv:5.1f} {flag:>3}")

print("─" * 75)
print(f"  Avg N: {sum(n60)/len(n60):.1f} | Avg Vs: {sum(vs)/len(vs):.1f} m/s")
print(f"  Weak: {len([n for n in n60 if n<15])} | Sat: {sum(saturated)} | Risk: {len(risky)} layers")
print("=" * 75)
````

## Complete Collection Comparison

| Feature | List `[]` | Tuple `()` | Set `{}` |
|---------|-----------|------------|----------|
| **Ordered** | Yes | Yes | No |
| **Mutable** | Yes | No | Yes (add/remove) |
| **Duplicates** | Allowed | Allowed | Not allowed |
| **Indexing** | `l[0]` | `t[0]` | Not possible |
| **Slicing** | `l[1:3]` | `t[1:3]` | Not possible |
| **Add items** | `.append()` | Cannot | `.add()` |
| **Create** | `[1, 2, 3]` | `(1, 2, 3)` | `{1, 2, 3}` |
| **From other** | `list(x)` | `tuple(x)` | `set(x)` |
| **Geotech use** | Field data, measurements | Coordinates, standards | Unique types, validation |

---

## Key Takeaways

1. **Lists** are ordered, mutable collections — the workhorse for data processing. Use `.append()` to build data incrementally, indexing/slicing to access subsets.

2. **Tuples** are immutable — they protect constants (coordinates, sieve sizes, standards). Use them when data should NEVER change after creation.

3. **Sets** store only unique values and support powerful comparison operations (union, intersection, difference). Perfect for finding distinct soil types and data validation.

4. **List comprehension** (`[f(x) for x in data if condition]`) replaces 4-line loops with 1 line. This is one of the most important Python features for data science.

5. You can **convert between types**: `list(my_set)`, `set(my_list)`, `tuple(my_list)` — choose the right container for each job.
