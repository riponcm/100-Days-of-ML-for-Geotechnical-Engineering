## Our Geotechnical Problem

Until now, our borehole data lived in **parallel lists**:

```python
depths     = [1.5, 3.0, 4.5, 6.0, ...]      # list 1
n_values   = [5,   8,   12,  18,  ...]      # list 2
soil_types = ["Fill", "Sand", "Sand", ...]   # list 3
```

The problem: to get Layer 4's data, you must know that **index 3** means the same thing across ALL lists. One wrong index = wrong data paired with wrong depth.

**Dictionaries** solve this by using **names** instead of positions:

```python
layer = {
    "depth": 6.0,
    "N60": 18,
    "soil": "Clayey Sand",
    "Vs": 240.4
}
# Access by NAME: layer["N60"] — no index guessing!
```

This mirrors how engineers actually think: a layer has a depth, an N-value, and a soil type — **named fields**, not numbered positions.

## Part 1: Creating Dictionaries — Key-Value Pairs

A **dictionary** maps **keys** (names) to **values** (data). Think of it like a labeled filing cabinet — each drawer has a label and contains one piece of information.

```python
# Syntax: {key: value, key: value, ...}
layer = {"depth": 6.0, "N60": 18, "soil": "Sand"}

#         ↑ key   ↑ value
```

**Rules:**
- Keys must be **immutable** (strings, numbers, tuples — NOT lists or dicts)
- Keys must be **unique** (each key appears only once)
- Values can be **anything**: strings, numbers, lists, other dicts
- Dictionaries are **ordered** (Python 3.7+) — insertion order is preserved
- Dictionaries are **mutable** — you can add, change, and delete entries

````python
# ─────────────────────────────────────────────────────
# CREATING DICTIONARIES — Labeled borehole data
# ─────────────────────────────────────────────────────

# One SPT layer as a dictionary
layer = {
    "depth": 6.0,
    "N60": 18,
    "soil": "Clayey Sand",
    "USCS": "SC",
    "saturated": True,
    "color": "grey"
}

print("Layer dictionary:")
print(layer)
print(f"Type:    {type(layer)}")
print(f"Entries: {len(layer)}")
print()

# Compare: Old way (parallel lists) vs New way (dictionary)
# OLD: depth=depths[3], N=n_values[3], soil=soils[3]  ← what does 3 mean?
# NEW: layer["depth"], layer["N60"], layer["soil"]    ← crystal clear!

print("Accessing by name:")
print(f"  Depth: {layer['depth']}m")
print(f"  N60:   {layer['N60']}")
print(f"  Soil:  {layer['soil']}")
print(f"  USCS:  {layer['USCS']}")
print(f"  Sat:   {layer['saturated']}")
````

````python
# ─────────────────────────────────────────────────────
# DIFFERENT WAYS TO CREATE DICTIONARIES
# ─────────────────────────────────────────────────────

# 1. Direct creation (most common)
site = {"name": "BH-01", "depth": 12.0, "wt": 3.2}
print(f"Direct: {site}")

# 2. Empty dict — fill later
record = {}
record["project"] = "Highway Bridge"
record["date"] = "2026-02-19"
print(f"Built:  {record}")

# 3. From two lists using zip() + dict()
keys = ["depth", "N60", "soil"]
values = [6.0, 18, "Sand"]
layer = dict(zip(keys, values))
print(f"From zip: {layer}")

# 4. Using dict() with keyword arguments
layer2 = dict(depth=3.0, N60=8, soil="Silty Sand")
print(f"Keywords: {layer2}")
print()
print("Method 1 (direct) and Method 3 (from zip) are most common.")
````

## Part 2: Accessing Values — Square Brackets vs `.get()`

Two ways to retrieve values from a dictionary:

| Method | Syntax | If Key Missing | When to Use |
|--------|--------|---------------|-------------|
| `[]` | `d["key"]` | **KeyError** (crash!) | When key MUST exist |
| `.get()` | `d.get("key", default)` | Returns `None` or default | When key MIGHT be missing |

In real data, fields are often missing. A driller might record N-value but forget the color. `.get()` handles this safely.

````python
# ─────────────────────────────────────────────────────
# ACCESSING VALUES — [] vs .get()
# ─────────────────────────────────────────────────────

layer = {"depth": 6.0, "N60": 18, "soil": "Clayey Sand", "Vs": 240.4}

# Method 1: Square brackets — direct access
print(f"layer['N60']  = {layer['N60']}")
print(f"layer['soil'] = {layer['soil']}")
# layer['color']  ← This would crash with KeyError!
print()

# Method 2: .get() — safe access with fallback
print(f"layer.get('N60')              = {layer.get('N60')}")
print(f"layer.get('color')            = {layer.get('color')}")       # None
print(f"layer.get('color', 'unknown') = {layer.get('color', 'unknown')}")  # default
print(f"layer.get('color', 'N/A')     = {layer.get('color', 'N/A')}")  
print()

# Practical: safely calculate with potentially missing data
gamma = layer.get('unit_weight', 18.0)  # use default if missing
sigma_v = gamma * layer['depth']
print(f"γ = {gamma} kN/m³ (default, since 'unit_weight' not in dict)")
print(f"σv = {sigma_v:.1f} kN/m²")
print()

# Check if a key exists
print(f"'N60' in layer:   {'N60' in layer}")
print(f"'color' in layer: {'color' in layer}")
````

## Part 3: Modifying Dictionaries — Add, Update, Delete

Dictionaries are **mutable** — you can change them after creation.

| Operation | Syntax | Example |
|-----------|--------|--------|
| Add new key | `d["new_key"] = value` | `layer["Vs"] = 240.4` |
| Update value | `d["key"] = new_value` | `layer["N60"] = 20` |
| Update multiple | `d.update({...})` | `layer.update({"Vs": 245, "Dr": 65})` |
| Delete key | `del d["key"]` | `del layer["color"]` |
| Remove & return | `d.pop("key")` | `vs = layer.pop("Vs")` |

````python
# ─────────────────────────────────────────────────────
# MODIFYING — Add, update, delete entries
# ─────────────────────────────────────────────────────

layer = {"depth": 6.0, "N60": 18, "soil": "Clayey Sand"}
print(f"Original:      {layer}")

# ADD new key-value pairs
layer["Vs"] = 240.4
layer["saturated"] = True
print(f"After add:     {layer}")

# UPDATE existing value (corrected field reading)
layer["N60"] = 20
print(f"After update:  N60 changed to {layer['N60']}")

# DELETE a key
del layer["saturated"]
print(f"After delete:  {layer}")

# POP — remove AND get the value back
vs = layer.pop("Vs")
print(f"Popped Vs={vs}: {layer}")
print()

# UPDATE multiple at once
layer.update({"Vs": 245.0, "Dr": 65.0, "color": "grey"})
print(f"After .update(): {layer}")
print()
print("Tip: .update() is great for batch corrections from the lab.")
````

## Part 4: Looping Through Dictionaries — keys, values, items

Three ways to loop through a dictionary:

| Method | What You Get | When to Use |
|--------|-------------|------------|
| `for key in d:` | Just the keys | When you only need key names |
| `for val in d.values():` | Just the values | When you don't need the keys |
| `for key, val in d.items():` | Both key AND value | **Most common** — you usually need both |

````python
# ─────────────────────────────────────────────────────
# LOOPING THROUGH DICTIONARIES
# ─────────────────────────────────────────────────────

layer = {
    "depth": 6.0,
    "N60": 18,
    "soil": "Clayey Sand",
    "Vs": 240.4,
    "Dr": 65.0
}

# .keys() — all key names
print(f"Keys:   {list(layer.keys())}")

# .values() — all values
print(f"Values: {list(layer.values())}")

# .items() — key-value pairs as tuples
print(f"Items:  {list(layer.items())}")
print()

# Most common: loop with .items()
print("Loop through key-value pairs:")
for key, value in layer.items():
    print(f"  {key:8s} = {value}")
print()

# Practical: build a formatted report from any dict
print("Formatted layer report:")
print("  " + "─" * 30)
for key, value in layer.items():
    if isinstance(value, float):
        print(f"  {key:8s} : {value:.1f}")
    else:
        print(f"  {key:8s} : {value}")
print("  " + "─" * 30)
````

## Part 5: Nested Dictionaries — Complex Borehole Records

Dictionaries can contain **other dictionaries** and **lists of dictionaries**. This lets you build complex, hierarchical data structures — like a complete borehole record with location info and multiple layers.

```
borehole
├── "id": "BH-01"
├── "location"
│   ├── "latitude": 40.7608
│   ├── "longitude": -111.8910
│   └── "city": "Salt Lake City"
├── "water_table": 3.2
└── "layers" (list)
    ├── {"depth": 1.5, "N60": 5, "soil": "Fill"}
    ├── {"depth": 3.0, "N60": 8, "soil": "Sand"}
    └── ...
```

Access nested data by **chaining** keys: `borehole["location"]["city"]`

````python
# ─────────────────────────────────────────────────────
# NESTED DICTIONARIES — Complete borehole record
# ─────────────────────────────────────────────────────

borehole = {
    "id": "BH-01",
    "location": {                            # Dict inside dict
        "latitude": 40.7608,
        "longitude": -111.8910,
        "elevation": 1320.5,
        "city": "Salt Lake City"
    },
    "water_table": 3.2,
    "date": "2026-02-19",
    "driller": "ABC Drilling Co.",
    "layers": [                              # List of dicts
        {"depth": 1.5, "N60": 5,  "soil": "Fill"},
        {"depth": 3.0, "N60": 8,  "soil": "Silty Sand"},
        {"depth": 4.5, "N60": 12, "soil": "Sand"},
        {"depth": 6.0, "N60": 18, "soil": "Clayey Sand"},
        {"depth": 7.5, "N60": 28, "soil": "Sand"},
    ]
}

# Access nested data — chain the keys
print(f"ID:   {borehole['id']}")
print(f"City: {borehole['location']['city']}")
print(f"Lat:  {borehole['location']['latitude']}")
print(f"WT:   {borehole['water_table']}m")
print()

# Access a specific layer (list index + dict key)
layer_3 = borehole["layers"][2]  # 3rd layer (index 2)
print(f"Layer 3: {layer_3}")
print(f"  Soil: {layer_3['soil']}")
print(f"  N60:  {layer_3['N60']}")
print()

# Loop through all layers
print("All layers:")
for i, layer in enumerate(borehole["layers"], start=1):
    print(f"  Layer {i}: {layer['depth']}m | N={layer['N60']:3d} | {layer['soil']}")
````

## Part 6: List of Dictionaries — The Most Common Data Pattern

A **list of dictionaries** is the single most important data pattern in Python data science:

```python
data = [
    {"depth": 1.5, "N60": 5,  "soil": "Fill"},      # Row 1
    {"depth": 3.0, "N60": 8,  "soil": "Sand"},      # Row 2
    {"depth": 6.0, "N60": 18, "soil": "Clay"},      # Row 3
]
```

- Each dictionary is a **row** (one record)
- The list is the **table** (all records)
- This is **exactly** how `pandas.DataFrame.to_dict('records')` works

To extract a "column": `[row["N60"] for row in data]`
To filter "rows": `[row for row in data if row["N60"] > 20]`

````python
# ─────────────────────────────────────────────────────
# LIST OF DICTIONARIES — Table-like data
# ─────────────────────────────────────────────────────

layers = [
    {"depth": 1.5,  "N60": 5,  "soil": "Fill",          "color": "dark brown"},
    {"depth": 3.0,  "N60": 8,  "soil": "Silty Sand",    "color": "grey-brown"},
    {"depth": 4.5,  "N60": 12, "soil": "Sand",           "color": "tan"},
    {"depth": 6.0,  "N60": 18, "soil": "Clayey Sand",    "color": "grey"},
    {"depth": 7.5,  "N60": 28, "soil": "Sand",           "color": "brown"},
    {"depth": 9.0,  "N60": 35, "soil": "Gravel",         "color": "grey"},
    {"depth": 10.5, "N60": 50, "soil": "Weathered Rock",  "color": "grey"},
    {"depth": 12.0, "N60": 60, "soil": "Bedrock",         "color": "grey"},
]

# Access a row by index, then field by name
print(f"Layer 4: depth={layers[3]['depth']}m, N={layers[3]['N60']}, soil={layers[3]['soil']}")
print()

# Extract a 'column' — list comprehension over dicts
all_n = [l["N60"] for l in layers]
all_depths = [l["depth"] for l in layers]
print(f"All N60:    {all_n}")
print(f"All depths: {all_depths}")
print()

# Filter rows by condition
sandy = [l for l in layers if "Sand" in l["soil"]]
print(f"Sandy layers ({len(sandy)}):")
for s in sandy:
    print(f"  {s['depth']}m | N={s['N60']} | {s['soil']}")
print()

# Print a formatted table
print(f"  {'#':>2} {'Depth':>6} {'N60':>4} {'Soil':16s} {'Color':12s}")
print("  " + "─" * 44)
for i, l in enumerate(layers, start=1):
    print(f"  {i:2d} {l['depth']:5.1f}m {l['N60']:4d} {l['soil']:16s} {l['color']:12s}")
````

## Part 7: Dictionary as Lookup Table — Instant Access by Key

One of the most powerful uses of dictionaries: **lookup tables**. Instead of searching through a list, you access data **directly** by key. This is O(1) — instant, no matter how big the dictionary is.

```python
# Without lookup (searching a list):
for i, soil in enumerate(soil_list):
    if soil == "Sand":
        gamma = weight_list[i]    # Slow! Searches every item.

# With lookup (dictionary):
gamma = soil_weight["Sand"]       # Instant! Direct access.
```

````python
# ─────────────────────────────────────────────────────
# LOOKUP TABLES — Soil properties by type
# ─────────────────────────────────────────────────────

# Unit weight lookup table
soil_unit_weight = {
    "Fill": 16.0,
    "Sand": 18.5,
    "Silty Sand": 17.5,
    "Clayey Sand": 18.0,
    "Clay": 17.0,
    "Silt": 16.5,
    "Gravel": 20.0,
    "Weathered Rock": 22.0,
    "Bedrock": 25.0,
}

# Instant lookup — no loops!
print(f"Sand:    γ = {soil_unit_weight['Sand']} kN/m³")
print(f"Gravel:  γ = {soil_unit_weight['Gravel']} kN/m³")
print(f"Bedrock: γ = {soil_unit_weight['Bedrock']} kN/m³")
print()

# Use in calculations — look up γ for each layer's soil type
layers_sample = [
    {"depth": 1.5, "soil": "Fill"},
    {"depth": 3.0, "soil": "Silty Sand"},
    {"depth": 6.0, "soil": "Sand"},
    {"depth": 9.0, "soil": "Gravel"},
]

print(f"  {'Depth':>5} {'Soil':16s} {'γ':>5} {'σv':>8}")
print("  " + "─" * 38)
for l in layers_sample:
    gamma = soil_unit_weight.get(l["soil"], 18.0)  # default 18 if unknown
    sigma_v = gamma * l["depth"]
    print(f"  {l['depth']:4.1f}m {l['soil']:16s} {gamma:5.1f} {sigma_v:7.1f}")
````

````python
# ─────────────────────────────────────────────────────
# USCS CLASSIFICATION LOOKUP
# ─────────────────────────────────────────────────────

uscs_description = {
    "GW": "Well-graded Gravel",
    "GP": "Poorly-graded Gravel",
    "GM": "Silty Gravel",
    "GC": "Clayey Gravel",
    "SW": "Well-graded Sand",
    "SP": "Poorly-graded Sand",
    "SM": "Silty Sand",
    "SC": "Clayey Sand",
    "ML": "Low-plasticity Silt",
    "CL": "Low-plasticity Clay",
    "MH": "High-plasticity Silt",
    "CH": "High-plasticity Clay",
    "OL": "Organic Silt/Clay (low)",
    "OH": "Organic Clay (high)",
    "PT": "Peat",
}

# Look up classifications
test_codes = ["SM", "CL", "SW", "CH", "XX"]
print("USCS Classification Lookup:")
for code in test_codes:
    desc = uscs_description.get(code, "Unknown — check code")
    print(f"  {code} → {desc}")

print()
print("Lookup tables replace if/elif chains with direct access.")
print("15 soil classifications in one dict vs 15 elif branches!")
````

## Part 9: Dictionary Comprehension — One-Line Dict Creation

Just like list comprehension (Day 005), but creates a **dictionary** instead of a list.

```python
# List comprehension  → [expression for var in iterable]
# Dict comprehension  → {key_expr: val_expr for var in iterable}
```

Example:

```python
# Traditional loop (4 lines):
depth_to_n = {}
for d, n in zip(depths, n_values):
    depth_to_n[d] = n

# Dict comprehension (1 line!):
depth_to_n = {d: n for d, n in zip(depths, n_values)}
```

````python
# ─────────────────────────────────────────────────────
# DICTIONARY COMPREHENSION — Build dicts in one line
# ─────────────────────────────────────────────────────

depths = [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]
n_values = [5, 8, 12, 18, 28, 35, 50, 60]

# Depth → N60 lookup
depth_to_n = {d: n for d, n in zip(depths, n_values)}
print(f"Depth → N: {depth_to_n}")
print(f"  N at 6.0m: {depth_to_n[6.0]}")
print()

# Depth → Vs (with calculation)
depth_to_vs = {d: round(97 * (n ** 0.314), 1)
               for d, n in zip(depths, n_values)}
print(f"Depth → Vs: {depth_to_vs}")
print()

# N → classification (with inline if/else)
n_to_class = {n: ("Dense+" if n >= 30 else "Medium" if n >= 10 else "Loose")
              for n in n_values}
print(f"N → Class: {n_to_class}")
print()

# With FILTER — only saturated layers
wt = 3.2
sat_n = {d: n for d, n in zip(depths, n_values) if d > wt}
print(f"Saturated only: {sat_n}")
````

````python
# ─────────────────────────────────────────────────────
# PRACTICAL DICT COMPREHENSION PATTERNS
# ─────────────────────────────────────────────────────

# Invert a dictionary (swap keys ↔ values)
soil_to_uscs = {"Sand": "SP", "Clay": "CL", "Silt": "ML"}
uscs_to_soil = {v: k for k, v in soil_to_uscs.items()}
print(f"Original:  {soil_to_uscs}")
print(f"Inverted:  {uscs_to_soil}")
print()

# Transform values (convert kPa to psf)
pressures_kpa = {"Layer 1": 27.75, "Layer 2": 55.5, "Layer 3": 111.0}
pressures_psf = {k: round(v * 20.89, 1) for k, v in pressures_kpa.items()}
print(f"kPa: {pressures_kpa}")
print(f"psf: {pressures_psf}")
print()

# Build summary from list of dicts
layers = [
    {"depth": 1.5, "N60": 5, "soil": "Fill"},
    {"depth": 3.0, "N60": 8, "soil": "Sand"},
    {"depth": 6.0, "N60": 18, "soil": "Clay"},
]
soil_at_depth = {l["depth"]: l["soil"] for l in layers}
print(f"Depth → Soil: {soil_at_depth}")
````

## Part 10: Counting & Grouping — Common Data Operations

Two everyday tasks made easy with dictionaries:

1. **Counting** — How many of each soil type? (Like SQL `GROUP BY COUNT`)
2. **Grouping** — Which layers are in each consistency class?

````python
# ─────────────────────────────────────────────────────
# COUNTING — Tally occurrences
# ─────────────────────────────────────────────────────

all_soils = ["Fill", "Sand", "Clay", "Sand", "Gravel",
             "Sand", "Clay", "Sand", "Bedrock"]

# Count each soil type using .get()
soil_count = {}
for soil in all_soils:
    soil_count[soil] = soil_count.get(soil, 0) + 1
#   ↑ key               ↑ current count (or 0)  ↑ add 1

print(f"Soil counts: {soil_count}")
print()

# Sort by count (most common first)
sorted_counts = sorted(soil_count.items(), key=lambda x: x[1], reverse=True)
print("Most common soils:")
for soil, count in sorted_counts:
    bar = "█" * count
    print(f"  {soil:10s} {count}  {bar}")
````

````python
# ─────────────────────────────────────────────────────
# GROUPING — Organize layers by category
# ─────────────────────────────────────────────────────

layers = [
    {"depth": 1.5,  "N60": 5,  "soil": "Fill"},
    {"depth": 3.0,  "N60": 8,  "soil": "Sand"},
    {"depth": 4.5,  "N60": 12, "soil": "Sand"},
    {"depth": 6.0,  "N60": 18, "soil": "Clay"},
    {"depth": 7.5,  "N60": 28, "soil": "Sand"},
    {"depth": 9.0,  "N60": 35, "soil": "Gravel"},
    {"depth": 10.5, "N60": 50, "soil": "Rock"},
    {"depth": 12.0, "N60": 60, "soil": "Bedrock"},
]

def classify(n):
    if n < 10: return "Loose"
    elif n < 30: return "Medium Dense"
    elif n < 50: return "Dense"
    else: return "Very Dense"

# Group layers by consistency class
groups = {}
for layer in layers:
    cls = classify(layer["N60"])
    if cls not in groups:
        groups[cls] = []           # Create empty list for new group
    groups[cls].append(layer)      # Add layer to its group

print("Layers grouped by consistency:")
for cls, group_layers in groups.items():
    depths = [l["depth"] for l in group_layers]
    n_vals = [l["N60"] for l in group_layers]
    print(f"  {cls:13s}: {len(group_layers)} layers")
    print(f"    Depths: {depths}")
    print(f"    N60:    {n_vals}")
````

````python
# ══════════════════════════════════════════════════════
# COMPLETE: SITE INVESTIGATION REPORT
# Using Dictionaries, Lookups, Comprehensions
# ══════════════════════════════════════════════════════

# Full site as structured dictionaries
site = {
    "project": "Highway Bridge Foundation",
    "location": "Salt Lake City, Utah",
    "coords": (40.7608, -111.8910),
    "boreholes": {
        "BH-01": {
            "wt": 3.2,
            "layers": [
                {"depth": 1.5,  "N60": 5,  "soil": "Fill"},
                {"depth": 3.0,  "N60": 8,  "soil": "Silty Sand"},
                {"depth": 4.5,  "N60": 12, "soil": "Sand"},
                {"depth": 6.0,  "N60": 18, "soil": "Clayey Sand"},
                {"depth": 7.5,  "N60": 28, "soil": "Sand"},
                {"depth": 9.0,  "N60": 35, "soil": "Gravel"},
                {"depth": 10.5, "N60": 50, "soil": "Weathered Rock"},
                {"depth": 12.0, "N60": 60, "soil": "Bedrock"},
            ]
        },
        "BH-02": {
            "wt": 2.8,
            "layers": [
                {"depth": 1.5,  "N60": 3,  "soil": "Fill"},
                {"depth": 3.0,  "N60": 6,  "soil": "Clay"},
                {"depth": 4.5,  "N60": 10, "soil": "Sand"},
                {"depth": 6.0,  "N60": 15, "soil": "Clayey Sand"},
                {"depth": 7.5,  "N60": 22, "soil": "Sand"},
                {"depth": 9.0,  "N60": 30, "soil": "Gravel"},
                {"depth": 10.5, "N60": 45, "soil": "Weathered Rock"},
                {"depth": 12.0, "N60": 55, "soil": "Bedrock"},
            ]
        },
    }
}

# Lookup table for unit weights
gamma_lookup = {
    "Fill": 16.0, "Sand": 18.5, "Silty Sand": 17.5,
    "Clayey Sand": 18.0, "Clay": 17.0, "Gravel": 20.0,
    "Weathered Rock": 22.0, "Bedrock": 25.0,
}

# Header
lat, lon = site["coords"]
print("=" * 75)
print(f"  {site['project']}")
print(f"  {site['location']} ({lat}, {lon})")
print(f"  Boreholes: {list(site['boreholes'].keys())}")
print("=" * 75)

# Process each borehole
bh_stats = {}  # Will store per-borehole summaries

for bh_id, bh in site["boreholes"].items():
    print(f"\n  {bh_id} | WT: {bh['wt']}m")
    print(f"  {'─' * 68}")
    print(f"  {'#':>3} {'Dep':>5} {'N60':>4} {'Soil':16s} {'Class':13s} {'Vs':>6} {'σv':>7}")
    print(f"  {'─' * 68}")
    
    n_all, vs_all = [], []
    
    for i, l in enumerate(bh["layers"], 1):
        Vs = 97 * (l["N60"] ** 0.314)
        gamma = gamma_lookup.get(l["soil"], 18.0)
        sigma_v = gamma * l["depth"]
        
        if l["N60"] < 4: cls = "Very Loose"
        elif l["N60"] < 10: cls = "Loose"
        elif l["N60"] < 30: cls = "Medium Dense"
        elif l["N60"] < 50: cls = "Dense"
        else: cls = "Very Dense"
        
        n_all.append(l["N60"])
        vs_all.append(Vs)
        
        print(f"  {i:3d} {l['depth']:4.1f}m {l['N60']:4d} {l['soil']:16s} "
              f"{cls:13s} {Vs:5.1f} {sigma_v:6.1f}")
    
    # Per-borehole summary (stored as dict)
    bh_stats[bh_id] = {
        "avg_n": sum(n_all) / len(n_all),
        "avg_vs": sum(vs_all) / len(vs_all),
        "min_n": min(n_all),
        "max_n": max(n_all),
        "weak": len([n for n in n_all if n < 15]),
        "soils": set(l["soil"] for l in bh["layers"]),
    }
    
    s = bh_stats[bh_id]
    print(f"  {'─' * 68}")
    print(f"  Avg N: {s['avg_n']:.1f} | Avg Vs: {s['avg_vs']:.1f} m/s | Weak: {s['weak']}")

# Cross-borehole comparison
print(f"\n{'═' * 75}")
print("  CROSS-BOREHOLE COMPARISON")
print(f"{'─' * 75}")
for bh_id, s in bh_stats.items():
    print(f"  {bh_id}: Avg N={s['avg_n']:.1f} | N range: {s['min_n']}–{s['max_n']} | "
          f"Unique soils: {len(s['soils'])}")

# Common soils (set intersection)
all_soil_sets = [s["soils"] for s in bh_stats.values()]
common = all_soil_sets[0] & all_soil_sets[1]
print(f"  Common soils: {common}")
print("=" * 75)
````

## Complete Dictionary Methods Reference

### Creating
| Method | Example |
|--------|--------|
| Direct | `{"depth": 6.0, "N60": 18}` |
| Empty + fill | `d = {}; d["key"] = val` |
| From zip | `dict(zip(keys, values))` |
| Comprehension | `{d: n for d, n in zip(depths, n_vals)}` |

### Accessing
| Method | Missing Key | Example |
|--------|------------|--------|
| `d[key]` | KeyError | `layer["N60"]` |
| `d.get(key)` | None | `layer.get("color")` |
| `d.get(key, default)` | default value | `layer.get("color", "N/A")` |
| `key in d` | — | `"N60" in layer` → True/False |

### Modifying
| Method | Example |
|--------|--------|
| Add/Update | `d["Vs"] = 240.4` |
| Update multiple | `d.update({"Vs": 245, "Dr": 65})` |
| Delete | `del d["color"]` |
| Pop (remove + return) | `vs = d.pop("Vs")` |

### Iterating
| Method | Returns | Example |
|--------|---------|--------|
| `d.keys()` | All keys | `list(layer.keys())` |
| `d.values()` | All values | `list(layer.values())` |
| `d.items()` | Key-value tuples | `for k, v in layer.items():` |

---

## Key Takeaways

1. **Dictionaries** store data as **key-value pairs** — access by name instead of position. `layer["N60"]` is clearer than `n_values[3]`.

2. **`.get(key, default)`** safely handles missing data — essential for real-world datasets where fields may be incomplete.

3. A **list of dictionaries** is the most important data pattern in Python data science — each dict is a row, the list is a table. This is exactly how Pandas DataFrames work.

4. **Lookup tables** replace long if/elif chains: put 15 USCS classifications in one dict instead of 15 elif branches.

5. **Dict comprehension** (`{k: v for k, v in data}`) creates dictionaries in one line — the dictionary version of list comprehension.
