## Part 1: Basic if / else — Is This Layer Saturated?

The simplest decision: **yes or no**.

Think of it like this:

```
IF the sample depth is deeper than the water table:
    → The soil is saturated (water fills the pores)
ELSE:
    → The soil is dry (above water table)
```

### Syntax

```python
if condition:
    # this runs when condition is True
else:
    # this runs when condition is False
```

**Important:** The colon `:` at the end of each line and the **indentation** (4 spaces) are mandatory. Python uses indentation to know what code belongs inside the if/else blocks.

````python
# ─────────────────────────────────────────────────────
# IS THIS LAYER SATURATED?
# ─────────────────────────────────────────────────────
# If the sample is deeper than the water table, it's saturated.
# This is a fundamental check in every borehole log.

water_table_depth = 3.2  # meters below ground surface
sample_depth = 6.5       # depth of our SPT sample

if sample_depth > water_table_depth:
    is_saturated = True
    print(f"Depth {sample_depth}m > Water table {water_table_depth}m")
    print("Result: Layer is SATURATED")
else:
    is_saturated = False
    print(f"Depth {sample_depth}m <= Water table {water_table_depth}m")
    print("Result: Layer is DRY")
````

````python
# ─────────────────────────────────────────────────────
# TRY IT: Change sample_depth to 2.0 and run again.
# What happens? The else block runs instead!
# ─────────────────────────────────────────────────────

sample_depth = 2.0  # now ABOVE water table

if sample_depth > water_table_depth:
    print(f"{sample_depth}m is BELOW water table → Saturated")
else:
    print(f"{sample_depth}m is ABOVE water table → Dry")
````

### What just happened?

Python evaluated the condition `sample_depth > water_table_depth`:

- First run: `6.5 > 3.2` → `True` → ran the `if` block
- Second run: `2.0 > 3.2` → `False` → ran the `else` block

The condition is always a **boolean** (`True` or `False`). Python only runs the code inside the matching block.

## Part 2: if / elif / else — Soil Consistency Classification

Often we don't have just two options. Soil consistency has **5 categories** based on SPT N60-value (for granular soils):

| N60 Range | Consistency | Settlement Risk |
|-----------|-------------|----------------|
| < 4 | Very Loose | HIGH |
| 4 – 9 | Loose | MODERATE |
| 10 – 29 | Medium Dense | LOW |
| 30 – 49 | Dense | VERY LOW |
| ≥ 50 | Very Dense | NEGLIGIBLE |

We need **elif** (short for "else if") to check multiple conditions in sequence:

```
IF N60 < 4      → Very Loose
ELIF N60 < 10   → Loose
ELIF N60 < 30   → Medium Dense
ELIF N60 < 50   → Dense
ELSE            → Very Dense
```

Python checks from top to bottom and **stops at the first True condition**.

````python
# ─────────────────────────────────────────────────────
# SOIL CONSISTENCY FROM SPT N60
# ─────────────────────────────────────────────────────
# This is straight from Terzaghi & Peck classification.
# Every geotechnical engineer knows this table by heart.
# Now Python knows it too.

N60 = 22.0

if N60 < 4:
    consistency = "Very Loose"
    risk_level = "HIGH"
elif N60 < 10:
    consistency = "Loose"
    risk_level = "MODERATE"
elif N60 < 30:
    consistency = "Medium Dense"
    risk_level = "LOW"
elif N60 < 50:
    consistency = "Dense"
    risk_level = "VERY LOW"
else:
    consistency = "Very Dense"
    risk_level = "NEGLIGIBLE"

print(f"N60 = {N60}")
print(f"Consistency: {consistency}")
print(f"Settlement Risk: {risk_level}")
````

````python
# ─────────────────────────────────────────────────────
# CLAY CONSISTENCY (different scale!)
# ─────────────────────────────────────────────────────
# For cohesive soils (clays), the N-value ranges are different.
# This shows how the SAME concept (elif chain) applies
# to different classification systems.

N60_clay = 6
soil_type = "Clay"

if soil_type in ("Sand", "Silty Sand", "Gravel"):
    # Granular soil classification
    if N60_clay < 4:
        consistency = "Very Loose"
    elif N60_clay < 10:
        consistency = "Loose"
    elif N60_clay < 30:
        consistency = "Medium Dense"
    elif N60_clay < 50:
        consistency = "Dense"
    else:
        consistency = "Very Dense"
elif soil_type in ("Clay", "Silty Clay", "Silt"):
    # Cohesive soil classification (different ranges!)
    if N60_clay < 2:
        consistency = "Very Soft"
    elif N60_clay < 4:
        consistency = "Soft"
    elif N60_clay < 8:
        consistency = "Medium Stiff"
    elif N60_clay < 15:
        consistency = "Stiff"
    elif N60_clay < 30:
        consistency = "Very Stiff"
    else:
        consistency = "Hard"
else:
    consistency = "Unknown (check soil type)"

print(f"Soil: {soil_type}, N60 = {N60_clay}")
print(f"Consistency: {consistency}")
print()
print("Notice: Same N60=6 gives 'Loose' for Sand but 'Medium Stiff' for Clay.")
print("The classification system depends on the soil type!")
````

### How elif Works — Step by Step

For `N60 = 22.0`, Python does this:

```
1. Is 22 < 4?   → No  → skip
2. Is 22 < 10?  → No  → skip
3. Is 22 < 30?  → YES → run this block → STOP (don't check rest)
4. Is 22 < 50?  → never checked
5. else          → never reached
```

**Key point:** Once Python finds a True condition, it runs that block and **skips everything below**. Order matters!

## Part 3: Comparison Operators — Engineering Checks

These are the building blocks of every condition:

| Operator | Meaning | Engineering Example |
|----------|---------|--------------------|
| `==` | Equal to | `soil == "Clay"` |
| `!=` | Not equal to | `status != "Failed"` |
| `>` | Greater than | `load > capacity` → overstressed! |
| `<` | Less than | `FoS < 1.5` → unsafe |
| `>=` | Greater than or equal | `N60 >= 50` → refusal |
| `<=` | Less than or equal | `depth <= water_table` → dry |

Each operator returns `True` or `False` — a boolean. That boolean is what `if` checks.

````python
# ─────────────────────────────────────────────────────
# BEARING CAPACITY SAFETY CHECK
# ─────────────────────────────────────────────────────
# Is the applied load safe for the soil's bearing capacity?

bearing_capacity = 150.0   # kN/m² (what the soil can handle)
applied_load = 180.0       # kN/m² (what the structure applies)

# Let's see what each comparison returns
print("Comparison Results:")
print(f"  load == capacity  : {applied_load == bearing_capacity}")
print(f"  load != capacity  : {applied_load != bearing_capacity}")
print(f"  load > capacity   : {applied_load > bearing_capacity}    ← This is the danger check!")
print(f"  load < capacity   : {applied_load < bearing_capacity}")
print(f"  load >= capacity  : {applied_load >= bearing_capacity}")
print(f"  load <= capacity  : {applied_load <= bearing_capacity}")
print()

# Now use it in a decision
if applied_load > bearing_capacity:
    FoS = bearing_capacity / applied_load
    print(f"OVERSTRESSED! FoS = {FoS:.2f} (< 1.0)")
    print("Action: Redesign foundation or improve soil")
elif applied_load > (bearing_capacity / 1.5):
    FoS = bearing_capacity / applied_load
    print(f"MARGINAL. FoS = {FoS:.2f} (between 1.0 and 1.5)")
    print("Action: Review design — consider increasing footing size")
else:
    FoS = bearing_capacity / applied_load
    print(f"SAFE. FoS = {FoS:.2f} (>= 1.5)")
    print("Action: Design is adequate")
````

### Common Mistake Alert

| Wrong | Right | Why |
|-------|-------|-----|
| `if load = 150:` | `if load == 150:` | Single `=` assigns, double `==` compares |
| `if FoS > 1.5 and < 3.0:` | `if FoS > 1.5 and FoS < 3.0:` | Must repeat the variable |

This is one of the most common beginner bugs. `=` means "set this value", `==` means "is this equal to?".

## Part 4: Logical Operators — Combining Conditions

Real engineering decisions rarely depend on just ONE condition. Liquefaction risk, for example, requires checking multiple factors simultaneously.

| Operator | Meaning | True When |
|----------|---------|----------|
| `and` | Both must be true | `is_saturated and N60 < 20` |
| `or` | At least one must be true | `N60 < 4 or is_saturated` |
| `not` | Flip true/false | `not is_saturated` → is it dry? |

### Truth Table (think of it like a decision matrix)

| A | B | A and B | A or B | not A |
|---|---|---------|--------|-------|
| True | True | **True** | True | False |
| True | False | False | **True** | False |
| False | True | False | **True** | True |
| False | False | False | False | **True** |

````python
# ─────────────────────────────────────────────────────
# LIQUEFACTION SCREENING with LOGICAL OPERATORS
# ─────────────────────────────────────────────────────
# Liquefaction requires ALL of these to be true:
#   1. Soil is saturated
#   2. Sandy soil (not clay)
#   3. Low N-value (< 20)
#   4. In a seismic zone
#   5. Shallow depth (< 20m)

# Layer data
N60 = 12
depth = 5.0
is_saturated = True
soil_type = "Sand"
seismic_zone = True

# Individual checks
is_sandy = soil_type in ("Sand", "Silty Sand", "Sandy Silt")
is_weak = N60 < 20
is_shallow = depth < 20.0

print("Individual checks:")
print(f"  Saturated?     {is_saturated}")
print(f"  Sandy soil?    {is_sandy}")
print(f"  Weak (N<20)?   {is_weak}")
print(f"  Shallow (<20m)?{is_shallow}")
print(f"  Seismic zone?  {seismic_zone}")
print()

# AND — ALL must be true for liquefaction risk
liq_risk = is_saturated and is_sandy and is_weak and is_shallow and seismic_zone
print(f"Liquefaction risk (all must be true): {liq_risk}")
print()

# Using it in a decision
if is_saturated and is_sandy and is_weak and seismic_zone:
    print("RESULT: Liquefaction analysis REQUIRED")
    print("Recommend: Seed & Idriss simplified procedure")
else:
    print("RESULT: Liquefaction risk is LOW")
````

````python
# ─────────────────────────────────────────────────────
# OR — When ANY problem is enough to flag a layer
# ─────────────────────────────────────────────────────

N60 = 3
depth = 1.0
is_organic = False

# A layer needs attention if ANY of these is true:
needs_attention = (N60 < 4) or (depth < 2.0) or is_organic

print(f"N60 < 4?     {N60 < 4}")
print(f"Depth < 2m?  {depth < 2.0}")
print(f"Organic?     {is_organic}")
print(f"Needs attention (any one true)? {needs_attention}")
print()

# NOT — Inverting a condition
is_saturated = True
is_dry = not is_saturated
print(f"is_saturated = {is_saturated}")
print(f"is_dry = not is_saturated = {is_dry}")
print()
print("'not' simply flips True to False and False to True.")
````

## Part 5: Nested if — Step-by-Step Liquefaction Screening

Sometimes we want to check conditions **one at a time**, like a decision tree. Each check narrows down our conclusion.

This is called **nesting** — putting one if inside another.

```
Is it a seismic zone?
  └─ Yes → Is the soil sandy?
              └─ Yes → Is it saturated?
                          └─ Yes → Is N-value low?
                                      └─ Yes → LIQUEFACTION RISK!
```

This mirrors how engineers actually think through the problem.

````python
# ─────────────────────────────────────────────────────
# NESTED if — LIQUEFACTION DECISION TREE
# ─────────────────────────────────────────────────────
# Each level adds one more check.
# Notice how the indentation shows the nesting level.

soil_type = "Sand"
depth = 6.5
N60 = 15.0
is_saturated = True
earthquake_zone = True

print("Liquefaction Screening — Decision Tree")
print("=" * 45)

if earthquake_zone:
    print("[1] Seismic zone? YES")
    
    if soil_type in ("Sand", "Silty Sand", "Sandy Silt"):
        print(f"[2] Liquefiable soil ({soil_type})? YES")
        
        if is_saturated:
            print("[3] Saturated? YES")
            
            if depth < 20.0:
                print(f"[4] Shallow ({depth}m < 20m)? YES")
                
                if N60 < 20:
                    print(f"[5] Low resistance (N60={N60} < 20)? YES")
                    print()
                    print(">>> LIQUEFACTION RISK: HIGH <<<")
                else:
                    print(f"[5] N60={N60} >= 20 → Adequate")
                    print(">>> Liquefaction risk: LOW")
            else:
                print(f"[4] Depth={depth}m >= 20m → Too deep")
                print(">>> Liquefaction unlikely")
        else:
            print("[3] Not saturated → No liquefaction")
    else:
        print(f"[2] {soil_type} → Not a liquefiable soil")
else:
    print("[1] Not in seismic zone → Skip liquefaction")
````

### Nested vs Flat — When to Use Which?

| Approach | When to Use | Example |
|----------|-------------|--------|
| **Flat** (`and` in one line) | Simple check, all conditions equal weight | `if sat and weak and seismic:` |
| **Nested** (if inside if) | Step-by-step screening, want to show reasoning | Liquefaction decision tree above |

Both give the same result. Nested is more readable when you want to **explain each step** (like in a report). Flat is shorter for quick checks.

## Part 6: Ternary Operator — Quick One-Line Decisions

For simple yes/no assignments, Python has a shortcut:

```python
result = "value_if_true" if condition else "value_if_false"
```

This is called the **ternary operator**. It's perfect for quick labels and status flags.

````python
# ─────────────────────────────────────────────────────
# TERNARY — Quick decisions in one line
# ─────────────────────────────────────────────────────

# Is the soil suitable for shallow foundation?
N60 = 35.0
status = "Suitable" if N60 >= 15 else "Not Suitable"
print(f"N60 = {N60} → Foundation: {status}")

# Saturation check
depth = 2.5
wt = 3.2
condition = "Saturated" if depth > wt else "Dry"
print(f"Depth={depth}m, WT={wt}m → {condition}")

# Factor of Safety classification
FoS = 2.3
safety = "SAFE" if FoS >= 1.5 else ("MARGINAL" if FoS >= 1.0 else "UNSAFE")
print(f"FoS = {FoS} → {safety}")

print()
print("Tip: Ternary is great for simple checks.")
print("For complex logic with 3+ outcomes, use regular if/elif/else.")
````

## Part 7: Complete Application — Multi-Layer Borehole Assessment

Now let's combine **everything** from today into a real-world borehole safety assessment.

We'll process 6 layers from BH-01 and for each layer:
1. Classify soil consistency (if/elif/else)
2. Check saturation (comparison)
3. Screen liquefaction risk (logical operators)
4. Check foundation suitability (ternary)
5. Generate a formatted report

This uses a `for` loop to go through each layer — we'll learn loops properly on Day 003, but the concept here is simple: **do the same checks for every layer**.

````python
# ══════════════════════════════════════════════════════
# COMPLETE BOREHOLE SAFETY ASSESSMENT
# ══════════════════════════════════════════════════════

# Borehole data — each layer is a dictionary (preview of Day 006)
# Don't worry about the syntax yet — focus on the if/elif/else logic

layers = [
    {"depth": 1.5,  "N60": 5,  "soil": "Fill",         "fines": 40},
    {"depth": 3.0,  "N60": 8,  "soil": "Silty Sand",   "fines": 25},
    {"depth": 6.0,  "N60": 18, "soil": "Sand",          "fines": 8},
    {"depth": 9.0,  "N60": 35, "soil": "Dense Sand",    "fines": 5},
    {"depth": 12.0, "N60": 50, "soil": "Gravel",        "fines": 3},
    {"depth": 15.0, "N60": 60, "soil": "Weathered Rock", "fines": 0},
]

wt = 3.2        # water table depth
seismic = True   # Salt Lake City is in a seismic zone
warnings = 0     # counter for flagged layers

print("=" * 80)
print("  BOREHOLE SAFETY ASSESSMENT — BH-01")
print("  Location: Salt Lake City, Utah | Water Table: 3.2m | Seismic Zone: Yes")
print("=" * 80)
print(f"  {'Depth':>6s} | {'N60':>3s} | {'Soil':16s} | {'Consistency':13s} | Sat | Liq     | Found")
print("─" * 80)

for layer in layers:
    d = layer["depth"]
    n = layer["N60"]
    s = layer["soil"]
    fc = layer["fines"]
    
    # ── Consistency classification (if/elif/else) ──
    if n < 4:
        cons = "Very Loose"
    elif n < 10:
        cons = "Loose"
    elif n < 30:
        cons = "Medium Dense"
    elif n < 50:
        cons = "Dense"
    else:
        cons = "Very Dense"
    
    # ── Saturation check (comparison) ──
    sat = "Yes" if d > wt else "No"
    
    # ── Liquefaction screening (logical operators) ──
    if (d > wt) and (n < 20) and seismic and (fc < 35) and (d < 20):
        liq = "CHECK"
        warnings += 1
    else:
        liq = "OK"
    
    # ── Foundation suitability (ternary) ──
    suitable = "Yes" if n >= 15 else "No"
    
    # ── Print row ──
    print(f"  {d:5.1f}m | {n:3d} | {s:16s} | {cons:13s} | {sat:3s} | {liq:7s} | {suitable}")

print("─" * 80)

# ── Summary ──
if warnings > 0:
    print(f"  {warnings} layer(s) flagged for liquefaction analysis!")
    print("  Recommend: Seed & Idriss simplified procedure")
else:
    print("  No liquefaction concerns identified")

# Check if there's a suitable bearing layer below 3m
found_bearing = False
for layer in layers:
    if layer["depth"] > 3.0 and layer["N60"] >= 15:
        found_bearing = True
        break

if found_bearing:
    print("  Suitable bearing layer found below 3.0m")
else:
    print("  No suitable bearing layer — consider deep foundation")

print("=" * 80)
````

## Key Takeaways

1. **`if/elif/else`** is how Python makes decisions. It checks conditions top-to-bottom and runs the **first True block only**.

2. **Comparison operators** (`>`, `<`, `==`, `!=`, `>=`, `<=`) produce `True`/`False`. They are the building blocks of every condition.

3. **Logical operators** (`and`, `or`, `not`) combine multiple conditions. `and` = all must be true, `or` = any one, `not` = flip.

4. **Ternary** (`x if condition else y`) is a one-line shortcut for simple decisions.

5. **Nested if** mimics step-by-step engineering reasoning (decision trees). This is exactly how ML classification trees work internally!
