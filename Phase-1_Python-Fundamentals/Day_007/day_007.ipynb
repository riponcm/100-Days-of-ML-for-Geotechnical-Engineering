## Our Geotechnical Problem

On Days 001-006 we kept writing the **same calculations over and over**:

```python
# Every time we needed Vs:
Vs = 97 * (N60 ** 0.314)       # copy-paste

# Every time we needed to classify:
if N60 < 4: cons = "Very Loose"  # copy-paste
elif N60 < 10: cons = "Loose"    # copy-paste
...10 more lines...              # copy-paste
```

What if we have 100 boreholes with 10 layers each? That's **1,000 copy-pastes** of the same code. And if we find a bug, we'd need to fix it in 1,000 places!

**Functions** solve this: **write the logic once, call it forever.**

### The Lab Machine Analogy

Think of a function like a **lab test machine**:

```
┌─────────────────────────────────┐
│         LAB MACHINE             │
│    (function definition)        │
│                                 │
│  INPUT (sample)  → PROCESS →  OUTPUT (report)  │
│  (parameters)      (body)      (return value)   │
└─────────────────────────────────┘

1. You FEED it a sample        → parameters (input)
2. The machine PROCESSES it    → function body (logic)
3. It RETURNS a report         → return value (output)
```

You build the machine **once**. Then you feed it **any** sample, **any** time.

## Part 1: Your First Function — The Basics

A function has two steps:

**Step 1: DEFINE** it (build the machine)
```python
def function_name():     # 'def' keyword + name + colon
    """What it does."""  # docstring (optional but recommended)
    code                  # indented body — the actual work
    code
```

**Step 2: CALL** it (use the machine)
```python
function_name()           # name + parentheses
```

**Key rule:** Defining a function does NOT run it. You must CALL it to execute the code inside.

````python
# ─────────────────────────────────────────────────────
# YOUR FIRST FUNCTION — No input, no output
# Just does a task when called.
# ─────────────────────────────────────────────────────

# STEP 1: DEFINE (build the machine)
def print_report_header():
    """Prints a standard borehole report header."""
    print("╔══════════════════════════════════════════╗")
    print("║    GEOTECHNICAL INVESTIGATION REPORT     ║")
    print("║    Standard Penetration Test (SPT)       ║")
    print("╚══════════════════════════════════════════╝")

# STEP 2: CALL (use the machine)
print_report_header()
print()

# Call it again — same result, zero code duplication!
print("Calling the same function again:")
print_report_header()
print()

# Note: defining the function didn't print anything.
# Only CALLING it (with the parentheses) runs the code inside.
````

## Part 2: Parameters — Giving Functions Input

Most functions need **input** to work with. Parameters are **placeholders** that receive values when the function is called.

```python
def function_name(parameter):   # DEFINE with parameter
    result = do_something(parameter)
    return result

answer = function_name(value)    # CALL with actual value
```

- **Parameter** = the sample slot on the machine (accepts anything)
- **Argument** = the actual sample you put in (specific value)

```
def classify_granular(N60):     ← N60 is the PARAMETER (slot)
    ...

classify_granular(22)           ← 22 is the ARGUMENT (actual value)
classify_granular(8)            ← 8 is a different argument
```

````python
# ─────────────────────────────────────────────────────
# FUNCTION WITH PARAMETER — Soil Classification
# ─────────────────────────────────────────────────────
# The same if/elif chain from Day 002, wrapped in a function.
# We wrote this logic ONCE. Now it works for ANY N-value.

def classify_granular(N60):
    """Classify granular soil consistency from SPT N60."""
    if N60 < 4:
        return "Very Loose"
    elif N60 < 10:
        return "Loose"
    elif N60 < 30:
        return "Medium Dense"
    elif N60 < 50:
        return "Dense"
    else:
        return "Very Dense"

# Classify ANY N-value with ONE line!
print(f"N60=3  → {classify_granular(3)}")
print(f"N60=8  → {classify_granular(8)}")
print(f"N60=22 → {classify_granular(22)}")
print(f"N60=45 → {classify_granular(45)}")
print(f"N60=60 → {classify_granular(60)}")
print()

# Now use it in a loop — process a whole borehole!
n_values = [3, 8, 18, 35, 55]
print("Classify all layers:")
for n in n_values:
    print(f"  N60={n:3d} → {classify_granular(n)}")

print()
print("Without functions: 5 layers × 10 lines = 50 lines")
print("With functions:    5 layers × 1 line   = 5 lines")
````

## Part 3: Return Values — Getting Results Back

`return` sends a value **back** to wherever the function was called. Without `return`, the function does work but gives you `None` back.

```python
def estimate_vs(N60):
    Vs = 97 * (N60 ** 0.314)
    return Vs            # ← Sends Vs back to the caller

result = estimate_vs(22) # ← result now holds the Vs value
print(result)            # 256.0
```

**Lab analogy:** You send the soil sample (parameter). The lab processes it (function body). The lab returns the test report (return value). If the lab forgets to send the report (no `return`), you get `None`.

````python
# ─────────────────────────────────────────────────────
# RETURN VALUES — Getting results from functions
# ─────────────────────────────────────────────────────

def estimate_vs(N60):
    """Estimate shear wave velocity (Imai & Tonouchi, 1982)."""
    Vs = 97 * (N60 ** 0.314)
    return Vs

def overburden_pressure(depth, unit_weight):
    """Calculate vertical overburden pressure: σv = γ × z."""
    return unit_weight * depth

# Call and STORE the returned value in a variable
Vs = estimate_vs(22)
sigma_v = overburden_pressure(6.5, 18.5)

print(f"Vs at N60=22:  {Vs:.1f} m/s")
print(f"σv at 6.5m:    {sigma_v:.1f} kN/m²")
print()

# Chain: use output of one function as input to another
print(f"Vs at N60=30: {estimate_vs(30):.1f} m/s")
print(f"Vs at N60=50: {estimate_vs(50):.1f} m/s")
print()

# IMPORTANT: What happens WITHOUT return?
def broken_vs(N60):
    Vs = 97 * (N60 ** 0.314)
    # Oops — no return!

result = broken_vs(22)
print(f"Without return: result = {result}  ← None! The value was lost.")
print("Always use 'return' when you need the result outside the function.")
````

## Part 4: Default Parameters — Sensible Assumptions

Default parameters have **pre-set values**. If you don't provide a value, the default is used. If you do provide one, your value overrides it.

Think of a **form with pre-filled fields**:

```
┌──────────────────────────────────┐
│  SPT Correction Form             │
│                                  │
│  N_field: [____]  (required!)    │
│  Efficiency: [60]  (pre-filled)  │
│  Diameter:   [100] (pre-filled)  │
└──────────────────────────────────┘
```

Most tests use 60% efficiency (safety hammer) and 100mm bore — so we pre-fill those. But you CAN change them.

```python
def simple_n60(N_field, efficiency=60, diameter=100):
#              ↑ required     ↑ optional (default=60)  ↑ optional (default=100)
```

**Rule:** Required parameters first, then optional with defaults.

````python
# ─────────────────────────────────────────────────────
# DEFAULT PARAMETERS — Pre-filled form fields
# ─────────────────────────────────────────────────────

def simple_n60(N_field, efficiency=60, diameter=100):
    """
    Simplified N60 correction with sensible defaults.
    efficiency=60 → safety hammer (most common)
    diameter=100  → standard borehole size
    """
    CE = efficiency / 60.0
    CB = 1.0 if diameter <= 115 else 1.05 if diameter <= 150 else 1.15
    return N_field * CE * CB

# Use ALL defaults (most common case — just provide N_field)
print(f"All defaults:         N60 = {simple_n60(22):.1f}")

# Override efficiency only (auto hammer = 80%)
print(f"Auto hammer (80%):    N60 = {simple_n60(22, efficiency=80):.1f}")

# Override diameter only (large bore)
print(f"Large bore (200mm):   N60 = {simple_n60(22, diameter=200):.1f}")

# Override both
print(f"Custom (75%, 150mm):  N60 = {simple_n60(22, efficiency=75, diameter=150):.1f}")
print()

# Tip: Use keyword arguments for clarity
# simple_n60(22, 80, 150)           ← works but unclear
# simple_n60(22, efficiency=80)     ← much clearer!
print("Tip: Use keyword arguments (efficiency=80) for readability.")
````

## Part 5: Multiple Return Values

Python functions can return **several values at once**. This is incredibly useful when one analysis produces multiple results.

```python
def analyze():
    return value1, value2, value3     # Return a tuple

a, b, c = analyze()                   # Unpack into variables
```

This is like a lab test that returns **multiple measurements** from one sample — you get all results at once.

````python
# ─────────────────────────────────────────────────────
# MULTIPLE RETURN VALUES — Full layer analysis
# ─────────────────────────────────────────────────────

def analyze_layer(N_field, depth, water_table, efficiency=60):
    """
    Complete SPT analysis for one layer.
    Returns: N60, Vs, consistency, is_saturated, liq_risk
    """
    N60 = N_field * (efficiency / 60)
    Vs = 97 * (N60 ** 0.314)
    
    if N60 < 4: cons = "Very Loose"
    elif N60 < 10: cons = "Loose"
    elif N60 < 30: cons = "Medium Dense"
    elif N60 < 50: cons = "Dense"
    else: cons = "Very Dense"
    
    sat = depth > water_table
    liq = "HIGH" if (sat and N60 < 15) else "MOD" if (sat and N60 < 25) else "LOW"
    
    return N60, Vs, cons, sat, liq

# Unpack all 5 results into separate variables
N60, Vs, cons, sat, liq = analyze_layer(22, 6.5, 3.2)

print(f"N60:          {N60:.1f}")
print(f"Vs:           {Vs:.1f} m/s")
print(f"Consistency:  {cons}")
print(f"Saturated:    {sat}")
print(f"Liquefaction: {liq}")
print()

# Use in a loop — process ALL layers!
depths = [1.5, 3.0, 4.5, 6.0, 7.5, 9.0]
n_raw  = [5,   8,   12,  18,  28,  35]
wt = 3.2

print(f"  {'#':>2} {'Dep':>5} {'N60':>5} {'Vs':>6} {'Class':13s} {'Sat':>3} Liq")
print("  " + "─" * 50)
for i, (d, n) in enumerate(zip(depths, n_raw), 1):
    N60, Vs, cons, sat, liq = analyze_layer(n, d, wt)
    print(f"  {i:2d} {d:4.1f}m {N60:5.1f} {Vs:5.1f} {cons:13s} {'Y' if sat else 'N':>3} {liq}")
````

## Part 6: Returning Dictionaries — Named Results

When a function returns **many values**, a dictionary keeps them organized. Instead of remembering the ORDER of 5+ values, you access by NAME.

```python
# Tuple return — must remember order:
N60, Vs, cons, sat, liq = analyze_layer(22, 6.5, 3.2)
# What's the 4th value again? Was it sat or liq? 🤔

# Dict return — access by name:
result = full_analysis(22, 6.5, 3.2)
result["Vs"]          # Crystal clear!
result["saturated"]   # No confusion!
```

````python
# ─────────────────────────────────────────────────────
# RETURNING DICTIONARIES — Labeled results
# ─────────────────────────────────────────────────────

def full_layer_analysis(N_field, depth, soil_type, water_table, efficiency=60):
    """Complete layer analysis returning a labeled dictionary."""
    N60 = N_field * (efficiency / 60)
    Vs = 97 * (N60 ** 0.314)
    Dr = 21 * (N60 ** 0.5)
    
    if N60 < 4: cons = "Very Loose"
    elif N60 < 10: cons = "Loose"
    elif N60 < 30: cons = "Medium Dense"
    elif N60 < 50: cons = "Dense"
    else: cons = "Very Dense"
    
    return {
        "depth": depth,
        "N60": round(N60, 1),
        "Vs": round(Vs, 1),
        "Dr": round(Dr, 1),
        "consistency": cons,
        "soil": soil_type,
        "saturated": depth > water_table,
    }

# Get complete analysis as a labeled dictionary
result = full_layer_analysis(22, 6.0, "Sand", 3.2)

print("Result dictionary:")
for key, value in result.items():
    print(f"  {key:13s} = {value}")
print()

# Access specific values by NAME
print(f"Vs = {result['Vs']} m/s")
print(f"Saturated? {result['saturated']}")
print(f"Classification: {result['consistency']}")
````

## Part 7: Docstrings — Documenting Your Functions

A **docstring** is a special string right after `def` that explains what the function does. It's not a comment — it's built-in documentation that Python can access with `help()`.

The **NumPy style** (used by scikit-learn, Pandas, and most ML libraries):

```python
def function_name(param1, param2):
    """
    One-line summary of what this function does.
    
    Parameters
    ----------
    param1 : type
        Description of param1
    param2 : type
        Description of param2
    
    Returns
    -------
    type
        Description of return value
    """
```

Your future self (and teammates) will thank you.

````python
# ─────────────────────────────────────────────────────
# DOCSTRINGS — Well-documented geotechnical function
# ─────────────────────────────────────────────────────

def bearing_capacity(N60, Df, B, foundation_shape="strip"):
    """
    Allowable bearing capacity from SPT (Meyerhof, 1965).
    
    Parameters
    ----------
    N60 : float
        Corrected SPT blow count
    Df : float
        Foundation depth in meters
    B : float
        Foundation width in meters
    foundation_shape : str, optional
        'strip' or 'square' (default: 'strip')
    
    Returns
    -------
    float
        Allowable bearing capacity in kPa (FoS=3)
    """
    factor = 12 if foundation_shape == "strip" else 8
    
    if Df / B <= 1:
        q_net = factor * N60 * (1 + 0.33 * Df / B)
    else:
        q_net = factor * N60 * 1.33
    
    return q_net / 3.0  # FoS = 3

# Access the documentation
help(bearing_capacity)

# Use it
q = bearing_capacity(22, 2.0, 2.0)
print(f"\nq_allow = {q:.1f} kPa")
````

## Part 8: Variable Scope — Where Variables Live

Variables created **inside** a function are **local** — they only exist during that function call. Once the function ends, they disappear.

Variables created **outside** functions are **global** — accessible everywhere.

```
Global scope (main program):
├── project = "Bridge"            ← accessible everywhere
│
├── def calculate(x):              ← function starts
│       factor = 0.1              ← LOCAL: born here
│       result = x * factor
│       return result             ← function ends, 'factor' dies
│
├── print(project)                ← works (global)
├── print(factor)                 ← ERROR! (local, doesn't exist here)
```

This is a **good thing**: functions don't pollute your main code with temporary variables.

````python
# ─────────────────────────────────────────────────────
# SCOPE — Local vs Global variables
# ─────────────────────────────────────────────────────

project = "Highway Bridge"  # GLOBAL — exists everywhere

def calculate_settlement(N60, load):
    """Variables inside here are LOCAL."""
    factor = 0.1         # LOCAL — born here, dies when function ends
    settlement = factor * load / N60
    return settlement    # Only the RETURNED value escapes

s = calculate_settlement(22, 150)

print(f"Settlement: {s:.2f} mm")
print(f"Project: {project}")  # GLOBAL → works fine
print()

# Uncomment to see the error:
# print(factor)  # NameError! 'factor' is LOCAL

print("LOCAL  = born inside function, dies when function ends")
print("GLOBAL = born outside, accessible everywhere")
print("Functions keep their internal variables private — that's good!")
````

## Part 9: Lambda — Quick One-Line Functions

For very **simple** operations, `lambda` creates a function in **one line**:

```python
# Regular function (3 lines)
def square(x):
    return x ** 2

# Lambda (1 line) — same thing!
square = lambda x: x ** 2
```

### When to use each:

| Use `def` when... | Use `lambda` when... |
|-------------------|---------------------|
| Multiple lines of logic | Single expression |
| if/elif/else chains | Simple calculation |
| Need docstring | Sorting key |
| Reusable with a clear name | Quick transformation |

**Rule of thumb:** If it fits in one line AND you can understand it at a glance → lambda. Otherwise → def.

````python
# ─────────────────────────────────────────────────────
# LAMBDA — One-line functions
# ─────────────────────────────────────────────────────

# Regular vs Lambda — same result:
def vs_regular(N60):
    return 97 * (N60 ** 0.314)

vs_lambda = lambda N60: 97 * (N60 ** 0.314)

print(f"Regular: Vs(22) = {vs_regular(22):.1f} m/s")
print(f"Lambda:  Vs(22) = {vs_lambda(22):.1f} m/s")
print()

# Useful lambda examples
overburden = lambda depth, gamma=18.5: gamma * depth
relative_density = lambda N60: 21 * (N60 ** 0.5)
is_saturated = lambda depth, wt: depth > wt

print(f"σv at 6m:       {overburden(6.0):.1f} kN/m²")
print(f"Dr at N60=22:   {relative_density(22):.1f}%")
print(f"Sat at 5m?      {is_saturated(5.0, 3.2)}")
````

````python
# ─────────────────────────────────────────────────────
# LAMBDA WITH SORTING — The most common lambda use
# ─────────────────────────────────────────────────────

# Sort a list of dicts by a specific field
layers = [
    {"soil": "Fill",   "N60": 5,  "depth": 1.5},
    {"soil": "Sand",   "N60": 22, "depth": 6.0},
    {"soil": "Clay",   "N60": 8,  "depth": 3.0},
    {"soil": "Gravel", "N60": 35, "depth": 9.0},
]

# Sort by N60 (weakest first)
by_n = sorted(layers, key=lambda d: d["N60"])
print("Sorted by N60 (weakest first):")
for l in by_n:
    print(f"  {l['soil']:8s} N60={l['N60']}")
print()

# Sort by depth (shallowest first)
by_depth = sorted(layers, key=lambda d: d["depth"])
print("Sorted by depth:")
for l in by_depth:
    print(f"  {l['depth']}m  {l['soil']}")
print()

# Sort by N60 descending (strongest first)
by_strength = sorted(layers, key=lambda d: d["N60"], reverse=True)
print("Sorted by strength (strongest first):")
for l in by_strength:
    print(f"  {l['soil']:8s} N60={l['N60']}")

print()
print("Lambda as sort key is THE most common lambda use in data science.")
````

## Part 10: Functions Calling Functions — Building Pipelines

Real engineering software is built by combining **small, focused functions**. Each function does ONE thing well. Then you chain them together into a **pipeline**.

```
Raw N → [correct_n] → N60 → [calc_vs] → Vs → [classify] → "Medium Dense"
                                ↓
                           [calc_overburden] → σv
```

This is exactly how ML pipelines work: preprocess → train → evaluate → predict.

````python
# ─────────────────────────────────────────────────────
# FUNCTION PIPELINE — Small functions chained together
# ─────────────────────────────────────────────────────

def correct_n(N_field, efficiency=60):
    """Step 1: Correct raw N for efficiency."""
    return N_field * (efficiency / 60)

def calc_vs(N60):
    """Step 2: Estimate Vs."""
    return 97 * (N60 ** 0.314)

def calc_overburden(depth, gamma=18.5):
    """Step 3: Overburden pressure."""
    return gamma * depth

def classify(N60):
    """Step 4: Classify."""
    if N60 < 4: return "Very Loose"
    elif N60 < 10: return "Loose"
    elif N60 < 30: return "Medium Dense"
    elif N60 < 50: return "Dense"
    return "Very Dense"

def site_class(Vs30):
    """Step 5: ASCE 7 Site Class."""
    if Vs30 > 1500: return "A"
    elif Vs30 > 760: return "B"
    elif Vs30 > 360: return "C"
    elif Vs30 > 180: return "D"
    return "E"

# Chain them: output of one → input of next
N60 = correct_n(22)          # 22.0
Vs = calc_vs(N60)            # 256.0
sigma_v = calc_overburden(6.0)  # 111.0
cons = classify(N60)         # "Medium Dense"

print(f"Pipeline: N_raw=22 → N60={N60:.0f} → Vs={Vs:.1f} → {cons} → σv={sigma_v:.1f}")
print()
print("Each function is simple. Together they form a complete analysis.")
print("This is exactly how ML pipelines work!")
````

## 🔨 Mini Project 1: SPT Borehole Processor

Let's build a real function that processes an **entire borehole** using our toolkit.

````python
# ══════════════════════════════════════════════════════
# MINI PROJECT 1: BOREHOLE PROCESSOR FUNCTION
# ══════════════════════════════════════════════════════

def process_borehole(depths, n_raw, soils, water_table, bh_id="BH-01"):
    """
    Process entire borehole: raw field data → analyzed report.
    Returns: (processed_layers, summary_dict)
    """
    processed = []
    for d, n, s in zip(depths, n_raw, soils):
        N60 = correct_n(n)
        Vs = calc_vs(N60)
        processed.append({
            "depth": d, "N60": N60, "Vs": round(Vs, 1),
            "sigma_v": round(calc_overburden(d), 1),
            "consistency": classify(N60),
            "soil": s, "saturated": d > water_table
        })
    
    all_vs = [l["Vs"] for l in processed]
    all_n = [l["N60"] for l in processed]
    avg_vs = sum(all_vs) / len(all_vs)
    
    summary = {
        "bh_id": bh_id, "layers": len(processed),
        "avg_n": round(sum(all_n)/len(all_n), 1),
        "avg_vs": round(avg_vs, 1),
        "site_class": site_class(avg_vs),
        "weak": len([n for n in all_n if n < 15]),
    }
    return processed, summary

# Process BH-01
depths = [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]
n_raw  = [5, 8, 12, 18, 28, 35, 50, 60]
soils  = ["Fill", "Silty Sand", "Sand", "Clayey Sand",
          "Sand", "Gravel", "Weathered Rock", "Bedrock"]

layers, summary = process_borehole(depths, n_raw, soils, 3.2)

print(f"  {summary['bh_id']}: {summary['layers']} layers | "
      f"Avg N={summary['avg_n']} | Avg Vs={summary['avg_vs']} m/s | "
      f"Site Class {summary['site_class']}")
print()
print(f"  {'#':>2} {'Dep':>5} {'N60':>5} {'Soil':16s} {'Class':13s} {'Vs':>6} Sat")
print("  " + "─" * 56)
for i, l in enumerate(layers, 1):
    print(f"  {i:2d} {l['depth']:4.1f}m {l['N60']:5.1f} {l['soil']:16s} "
          f"{l['consistency']:13s} {l['Vs']:5.1f} {'Y' if l['saturated'] else 'N':>3}")
````

## 🔨 Mini Project 2: Bearing Capacity Calculator

````python
# ══════════════════════════════════════════════════════
# MINI PROJECT 2: BEARING CAPACITY COMPARISON
# ══════════════════════════════════════════════════════

def bearing_meyerhof(N60, Df, B, shape="strip"):
    """Allowable bearing capacity (Meyerhof, 1965), FoS=3."""
    factor = 12 if shape == "strip" else 8
    q_net = factor * N60 * min(1.33, 1 + 0.33 * Df / B)
    return round(q_net / 3.0, 1)

def safety_check(load, capacity):
    """Check factor of safety and return (FoS, status)."""
    FoS = capacity / load if load > 0 else float('inf')
    if FoS < 1.0: return round(FoS, 2), "FAILURE ✗"
    elif FoS < 1.5: return round(FoS, 2), "MARGINAL ⚠"
    elif FoS < 3.0: return round(FoS, 2), "SAFE ✓"
    return round(FoS, 2), "OVER-DESIGNED"

# Compare design scenarios
scenarios = [
    {"N60": 15, "Df": 1.5, "B": 2.0, "shape": "strip",  "load": 100},
    {"N60": 15, "Df": 1.5, "B": 2.0, "shape": "square", "load": 100},
    {"N60": 25, "Df": 2.0, "B": 3.0, "shape": "strip",  "load": 150},
    {"N60": 35, "Df": 2.0, "B": 2.5, "shape": "square", "load": 200},
]

print(f"  {'N60':>3} {'Df':>4} {'B':>4} {'Shape':>6} {'q_allow':>8} {'Load':>5} {'FoS':>5} Status")
print("  " + "─" * 52)
for s in scenarios:
    q = bearing_meyerhof(s["N60"], s["Df"], s["B"], s["shape"])
    fos, status = safety_check(s["load"], q)
    print(f"  {s['N60']:3d} {s['Df']:3.1f}m {s['B']:3.1f}m {s['shape']:>6} "
          f"{q:7.1f} {s['load']:5d} {fos:5.2f} {status}")
````

## 🔨 Mini Project 3: Multi-Borehole Site Comparator

````python
# ══════════════════════════════════════════════════════
# MINI PROJECT 3: COMPARE 3 BOREHOLES
# ══════════════════════════════════════════════════════

# Define borehole data
boreholes = {
    "BH-01": {"depths": [1.5, 3, 6, 9, 12], "n": [5, 8, 18, 35, 60],
              "soils": ["Fill", "Sand", "Sand", "Gravel", "Bedrock"], "wt": 3.2},
    "BH-02": {"depths": [1.5, 3, 6, 9, 12], "n": [3, 6, 15, 28, 55],
              "soils": ["Fill", "Clay", "Sand", "Gravel", "Bedrock"], "wt": 2.8},
    "BH-03": {"depths": [1.5, 3, 6, 9, 12], "n": [7, 12, 25, 40, 65],
              "soils": ["Fill", "Sand", "Sand", "Gravel", "Bedrock"], "wt": 4.0},
}

# Process each borehole with our function
summaries = {}
for bh_id, data in boreholes.items():
    _, summary = process_borehole(
        data["depths"], data["n"], data["soils"], data["wt"], bh_id
    )
    summaries[bh_id] = summary

# Comparison table
print("=" * 55)
print("  MULTI-BOREHOLE SITE COMPARISON")
print("=" * 55)
print(f"  {'Metric':<14} {'BH-01':>8} {'BH-02':>8} {'BH-03':>8} {'Best':>8}")
print("  " + "─" * 48)

for key, label in [("avg_n", "Avg N"), ("avg_vs", "Avg Vs"),
                    ("site_class", "Site Class"), ("weak", "Weak layers")]:
    vals = {bh: s[key] for bh, s in summaries.items()}
    
    if isinstance(list(vals.values())[0], (int, float)):
        if key == "weak":
            best = min(vals, key=vals.get)  # fewer weak = better
        else:
            best = max(vals, key=vals.get)  # higher = better
    else:
        best = "—"
    
    v = list(vals.values())
    print(f"  {label:<14} {str(v[0]):>8} {str(v[1]):>8} {str(v[2]):>8} {best:>8}")

print("=" * 55)
print("\nAll 3 boreholes processed with ONE reusable function!")
````

## Complete Functions Reference

### Defining & Calling
| Concept | Syntax | Example |
|---------|--------|--------|
| Define | `def name(params):` | `def calc_vs(N60):` |
| Call | `name(args)` | `calc_vs(22)` |
| Return | `return value` | `return 256.0` |
| Multi-return | `return a, b, c` | `return N60, Vs, cons` |
| Return dict | `return {"key": val}` | `return {"Vs": 256.0}` |

### Parameters
| Type | Syntax | Example |
|------|--------|--------|
| Required | `def f(x):` | Must provide |
| Default | `def f(x, y=10):` | Optional, has fallback |
| Keyword | `f(x=5, y=10)` | Named for clarity |

### Lambda
| Concept | Syntax | Example |
|---------|--------|--------|
| Basic | `lambda x: expr` | `lambda n: 97*n**0.314` |
| With default | `lambda x, y=18.5: x*y` | Overburden |
| Sort key | `sorted(data, key=lambda d: d["N60"])` | Sort dicts |

---

## Key Takeaways

1. **Functions = reusability.** Write the classification logic once, call it 1,000 times. `classify_granular(22)` replaces 10 lines of if/elif.

2. **Default parameters** set sensible assumptions (`efficiency=60`) that users can override when needed. Required params first, optional with defaults after.

3. **Return dictionaries** for complex results — `result["Vs"]` is clearer than remembering that Vs was the 3rd return value.

4. **Lambda** is for simple one-liners — especially as sort keys: `sorted(data, key=lambda d: d["N60"])`. For complex logic, always use `def`.

5. **Function pipelines** (small functions chained together) are how real engineering software and ML pipelines are built. Each function does ONE thing well.

## How This Connects to ML

| Today's Concept | ML Equivalent |
|----------------|---------------|

| Functions calling functions | `fit()` calls `forward()` calls `loss()` |
