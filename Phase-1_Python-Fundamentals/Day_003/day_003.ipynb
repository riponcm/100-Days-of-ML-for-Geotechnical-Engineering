## Part 1: Basic for Loop — Processing Every Layer

A `for` loop takes a collection (like a list) and runs code **once for each item**.

```python
for item in collection:
    # do something with item
```

Think of it like going through a borehole log sheet row by row:

```
FOR each layer IN the borehole log:
    read the N-value
    classify the soil
    print the result
```

````python
# ─────────────────────────────────────────────────────
# BASIC for LOOP — print every N-value
# ─────────────────────────────────────────────────────
# N-values from 8 layers of BH-01

n_values = [5, 8, 12, 18, 28, 35, 50, 60]

print("SPT N-values from BH-01:")
for n in n_values:
    print(f"  N = {n}")

# What happened:
# Iteration 1: n = 5  → prints "N = 5"
# Iteration 2: n = 8  → prints "N = 8"
# ... and so on for all 8 values
````

````python
# ─────────────────────────────────────────────────────
# LOOP + CONDITIONAL — Classify every layer
# ─────────────────────────────────────────────────────
# Combining Day 002 (if/elif/else) with Day 003 (loops).
# Write the classification ONCE, apply to ALL layers.

n_values = [5, 8, 12, 18, 28, 35, 50, 60]

print("Soil Consistency Classification:")
print(f"  {'N60':>4s}  {'Consistency':15s}")
print("  " + "─" * 25)

for n in n_values:
    if n < 4:
        cons = "Very Loose"
    elif n < 10:
        cons = "Loose"
    elif n < 30:
        cons = "Medium Dense"
    elif n < 50:
        cons = "Dense"
    else:
        cons = "Very Dense"
    
    print(f"  {n:4d}  {cons}")
````

### What Just Happened?

The loop ran **8 times** — once for each value in the list. Each time:

1. Python assigned the next value to `n`
2. The if/elif/else chain classified it
3. The result was printed

We wrote the classification logic **once**. The loop applied it to all 8 layers. If we had 800 layers, the same 10 lines of code would still work.

## Part 2: range() — Generating Number Sequences

`range()` creates a sequence of numbers. Very useful for:
- Generating depth intervals
- Running a loop a specific number of times
- Accessing list items by index

| Syntax | Produces | Example |
|--------|---------|--------|
| `range(n)` | 0, 1, 2, ... n-1 | `range(5)` → 0,1,2,3,4 |
| `range(start, stop)` | start, start+1, ... stop-1 | `range(2, 6)` → 2,3,4,5 |
| `range(start, stop, step)` | start, start+step, ... | `range(0, 10, 2)` → 0,2,4,6,8 |

````python
# ─────────────────────────────────────────────────────
# range() — Generate SPT sampling depths
# ─────────────────────────────────────────────────────
# SPT is typically done at 1.5m intervals.
# We need depths: 1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0

print("SPT sampling depths (1.5m intervals):")
for i in range(8):
    depth = 1.5 * (i + 1)    # i goes 0,1,2,...7 → depth goes 1.5,3.0,...12.0
    print(f"  Sample {i+1}: {depth:.1f} m")

print()

# range(start, stop, step) — every other layer
print("Every other layer (step=2):")
n_values = [5, 8, 12, 18, 28, 35, 50, 60]
for i in range(0, 8, 2):     # 0, 2, 4, 6
    print(f"  Index {i}: N = {n_values[i]}")

print()

# Counting down (like drilling depth increasing)
print("Countdown to drilling start:")
for i in range(3, 0, -1):    # 3, 2, 1
    print(f"  {i}...")
print("  Drilling started!")
````

````python
# ─────────────────────────────────────────────────────
# PRACTICAL: Count layers that meet a condition
# ─────────────────────────────────────────────────────
# How many layers have N > 20? (suitable for shallow foundation)

n_values = [5, 8, 12, 18, 28, 35, 50, 60]
threshold = 20

strong_count = 0
weak_count = 0

for n in n_values:
    if n >= threshold:
        strong_count += 1    # += means "add 1 to current value"
    else:
        weak_count += 1

print(f"Threshold: N >= {threshold}")
print(f"Strong layers: {strong_count}")
print(f"Weak layers  : {weak_count}")
print(f"Total layers : {len(n_values)}")
````

### The Accumulator Pattern

Notice the pattern above:

```python
counter = 0           # 1. Initialize BEFORE the loop
for item in data:     # 2. Loop through data
    if condition:      # 3. Check condition
        counter += 1  # 4. Accumulate (add to counter)
```

This **accumulator pattern** is one of the most common patterns in programming. You'll use it to compute sums, averages, mins, maxes, and counts — all fundamental to data analysis.

## Part 3: while Loop — Keep Going Until a Condition is Met

A `while` loop repeats **as long as a condition is True**. It stops the moment the condition becomes False.

```python
while condition:
    # do something
    # (something must eventually make condition False, or loop runs forever!)
```

### When to use `while` vs `for`?

| Use `for` when... | Use `while` when... |
|-------------------|--------------------|
| You know how many times to loop | You don't know when to stop |
| Iterating through a list | Waiting for a condition |
| Processing every item | Searching for something |
| *"Do this for each layer"* | *"Keep drilling until refusal"* |

````python
# ─────────────────────────────────────────────────────
# WHILE LOOP — Drilling until refusal
# ─────────────────────────────────────────────────────
# In real drilling, you don't know how deep you'll go.
# You keep drilling until:
#   - You hit refusal (N >= 50), or
#   - You reach maximum depth (budget/equipment limit)
#
# This is a perfect case for 'while' — we don't know
# how many iterations in advance.

current_depth = 0.0
increment = 1.5       # drill 1.5m per round
max_depth = 30.0
layer = 0

# Simulated N-values (imagine these come from field in real-time)
simulated_n = [3, 5, 8, 12, 18, 22, 28, 35, 42, 48, 50]

print("Drilling Log:")
print(f"  {'Depth':>8s}  {'N-value':>7s}  Action")
print("  " + "─" * 42)

while current_depth < max_depth:
    current_depth += increment
    
    # Get N-value for this depth
    if layer < len(simulated_n):
        n = simulated_n[layer]
    else:
        n = 50  # assume refusal
    
    if n >= 50:
        print(f"  {current_depth:7.1f}m  {n:>7d}  REFUSAL! Stop drilling.")
        break  # Exit the loop immediately
    else:
        print(f"  {current_depth:7.1f}m  {n:>7d}  Continue...")
    
    layer += 1

print(f"\nFinal depth: {current_depth:.1f} m")
print(f"Layers sampled: {layer + 1}")
````

````python
# ─────────────────────────────────────────────────────
# WHILE for iterative calculations
# ─────────────────────────────────────────────────────
# Iterative settlement calculation:
# Keep adding layer settlements until total exceeds
# the allowable settlement (25mm).

layer_settlements = [3.2, 5.1, 6.8, 4.5, 2.1, 1.3, 0.8, 0.3]  # mm per layer
allowable = 25.0  # mm

total_settlement = 0.0
layer_idx = 0

print("Cumulative Settlement Check:")
while total_settlement < allowable and layer_idx < len(layer_settlements):
    total_settlement += layer_settlements[layer_idx]
    status = "OK" if total_settlement < allowable else "EXCEEDED!"
    print(f"  After layer {layer_idx+1}: {total_settlement:.1f} mm [{status}]")
    layer_idx += 1

if total_settlement >= allowable:
    print(f"\nSettlement exceeds {allowable}mm at layer {layer_idx}!")
    print("Action: Consider ground improvement")
else:
    print(f"\nTotal settlement: {total_settlement:.1f}mm < {allowable}mm — OK")
````

### Infinite Loop Warning

A `while` loop runs **forever** if the condition never becomes False:

```python
# DANGER — this never stops!
x = 1
while x > 0:
    print(x)     # x never changes, so x > 0 is always True
```

**Always make sure** something inside the loop changes the condition (like `layer += 1` or `break`).

If you accidentally create an infinite loop in Jupyter, click the **stop button** (square icon) or press **Ctrl+C** in terminal.

## Part 4: enumerate() — Index + Value Together

Often you need **both** the position (index) and the value. For example: "Layer **3** is **Sand**".

```python
# Without enumerate (clunky):
for i in range(len(soils)):
    print(f"Layer {i}: {soils[i]}")

# With enumerate (clean):
for i, soil in enumerate(soils):
    print(f"Layer {i}: {soil}")
```

`enumerate()` gives you both the **counter** and the **value** in each iteration. Much cleaner!

````python
# ─────────────────────────────────────────────────────
# enumerate() — Numbered borehole log
# ─────────────────────────────────────────────────────

soil_descriptions = [
    "Fill (loose, dark brown)",
    "Silty Sand (SM, grey-brown)",
    "Sand (SP, medium dense, tan)",
    "Clayey Sand (SC, stiff, grey)",
    "Sand (SW, dense, brown)",
    "Dense Gravel (GP, grey)",
    "Weathered Sandstone",
    "Bedrock (refusal)",
]

# Default: index starts at 0
print("Borehole Log (index from 0):")
for index, description in enumerate(soil_descriptions):
    depth = 1.5 * (index + 1)
    print(f"  [{index}] {depth:5.1f}m — {description}")

print()
````

````python
# ─────────────────────────────────────────────────────
# enumerate(start=1) — Human-friendly numbering
# ─────────────────────────────────────────────────────
# Engineers count from 1, not 0!

print("Borehole Log (layer numbers from 1):")
for num, desc in enumerate(soil_descriptions, start=1):
    depth = 1.5 * num
    print(f"  Layer {num}: {depth:5.1f}m — {desc}")

print()
print("Tip: 'start=1' is perfect for engineering reports.")
print("     Use default (start=0) when accessing list indices.")
````

## Part 5: zip() — Combining Parallel Data

In geotechnical work, data often lives in **separate lists**:

```
depths:      [1.5,  3.0,  4.5,  6.0  ]
n_values:    [5,    8,    12,   18   ]
soil_types:  ["Fill", "Silty Sand", "Sand", "Clayey Sand"]
```

Each position represents the **same layer**. `zip()` pairs them together:

```
zip result:  (1.5, 5, "Fill"), (3.0, 8, "Silty Sand"), ...
```

Think of a **zipper** on a jacket — it interlocks two sides tooth by tooth.

````python
# ─────────────────────────────────────────────────────
# zip() — Pair depth + N-value + soil type
# ─────────────────────────────────────────────────────

depths =     [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]
n_values =   [5,   8,   12,  18,  28,  35,  50,   60]
soil_types = ["Fill", "Silty Sand", "Sand", "Clayey Sand",
              "Sand", "Gravel", "Weathered Rock", "Bedrock"]

print("Combined Borehole Log (3 lists zipped):")
print(f"  {'Depth':>6s}  {'N60':>4s}  {'Soil Type'}")
print("  " + "─" * 38)

for d, n, s in zip(depths, n_values, soil_types):
    print(f"  {d:5.1f}m  {n:4d}  {s}")

print()
print("Each iteration unpacks one item from EACH list simultaneously.")
print("d=1.5, n=5, s='Fill' → d=3.0, n=8, s='Silty Sand' → ...")
````

````python
# ─────────────────────────────────────────────────────
# THE ULTIMATE COMBO: enumerate + zip
# ─────────────────────────────────────────────────────
# Get layer number + depth + N-value + soil type all at once.
# This is the most common pattern in data processing.

print("Full Borehole Log (enumerate + zip):")
print(f"  {'#':>3s}  {'Depth':>6s}  {'N60':>4s}  {'Soil Type'}")
print("  " + "─" * 42)

for i, (d, n, s) in enumerate(zip(depths, n_values, soil_types), start=1):
    print(f"  {i:3d}  {d:5.1f}m  {n:4d}  {s}")

print()
print("Syntax breakdown:")
print("  enumerate(...)      → gives index i")
print("  zip(d, n, s)        → pairs the three lists")
print("  (d, n, s)           → unpacks each trio")
print("  start=1             → layer numbers from 1")
````

````python
# ─────────────────────────────────────────────────────
# zip() for CALCULATIONS across parallel lists
# ─────────────────────────────────────────────────────
# Calculate overburden pressure at each depth:
# σv = γ × depth

depths =      [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]
unit_weights = [16.0, 17.5, 18.0, 18.5, 19.0, 19.5, 20.0, 21.0]  # kN/m³

print("Overburden Pressure at Each Depth:")
print(f"  {'Depth':>6s}  {'γ (kN/m³)':>10s}  {'σv (kN/m²)':>10s}")
print("  " + "─" * 32)

for d, gamma in zip(depths, unit_weights):
    sigma_v = gamma * d
    print(f"  {d:5.1f}m  {gamma:10.1f}  {sigma_v:10.1f}")
````

## Part 6: Loop Control — break, continue, pass

These keywords give you fine control over what happens inside a loop:

| Keyword | What It Does | Geotechnical Analogy |
|---------|-------------|---------------------|
| `break` | Stop the loop immediately | Stop drilling at refusal |
| `continue` | Skip this iteration, go to next | Skip clay layers (only check sand) |
| `pass` | Do nothing (placeholder) | TODO: add calculation later |

````python
# ─────────────────────────────────────────────────────
# break — Find the FIRST suitable bearing layer
# ─────────────────────────────────────────────────────
# We only need the first layer with N >= 15.
# Once found, stop searching.

depths =     [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]
n_values =   [5,   8,   12,  18,  28,  35,  50,   60]
soil_types = ["Fill", "Silty Sand", "Sand", "Clayey Sand",
              "Sand", "Gravel", "Weathered Rock", "Bedrock"]

print("Searching for first bearing layer (N >= 15):")

for d, n, s in zip(depths, n_values, soil_types):
    if n >= 15:
        print(f"  FOUND at {d:.1f}m: N={n}, {s}")
        print(f"  Recommend: Place foundation at {d:.1f}m depth")
        break          # Stop! We found what we need.
    else:
        print(f"  {d:.1f}m: N={n} — too weak, keep searching...")

print()
print("Notice: Loop stopped at 6.0m. It never checked 7.5m or beyond.")
print("'break' saves time by stopping early.")
````

````python
# ─────────────────────────────────────────────────────
# continue — Skip layers that don't need liquefaction check
# ─────────────────────────────────────────────────────
# Only sandy soils can liquefy. Skip everything else.

wt = 3.2

print("Liquefaction Screening (sandy layers only):")

for d, n, s in zip(depths, n_values, soil_types):
    # Skip non-sandy soils — they don't liquefy
    if "Sand" not in s:
        continue       # Jump to next iteration immediately
    
    # Only sandy layers reach this point
    if d > wt and n < 20:
        print(f"  {d:5.1f}m | N={n:3d} | {s:16s} → LIQUEFIABLE")
    else:
        print(f"  {d:5.1f}m | N={n:3d} | {s:16s} → OK")

print()
print("Notice: Fill, Gravel, Weathered Rock, Bedrock were skipped entirely.")
print("'continue' skips the rest of the loop body and moves to the next item.")
````

````python
# ─────────────────────────────────────────────────────
# pass — Placeholder for future code
# ─────────────────────────────────────────────────────
# You know a calculation is needed but haven't written it yet.
# 'pass' lets the code run without errors.

print("Layers needing detailed analysis:")

for d, n in zip(depths, n_values):
    if n < 10:
        pass  # TODO: add Schmertmann settlement calculation
        print(f"  {d:.1f}m (N={n}) — settlement calc needed [placeholder]")
    elif n >= 50:
        pass  # TODO: add rock quality assessment
        print(f"  {d:.1f}m (N={n}) — rock quality check [placeholder]")

print()
print("'pass' does nothing — it's a placeholder so Python doesn't")
print("complain about an empty if/for/while block.")
````

## Part 7: Nested Loops — Processing Multiple Boreholes

A real site has **multiple boreholes**. We need a loop for each borehole, and inside it, a loop for each layer.

```
FOR each borehole:
    FOR each layer in that borehole:
        process the layer
```

This is called a **nested loop** — a loop inside a loop.

````python
# ─────────────────────────────────────────────────────
# NESTED LOOPS — Multi-borehole comparison
# ─────────────────────────────────────────────────────

boreholes = {
    "BH-01": [5, 8, 18, 35, 50],
    "BH-02": [3, 6, 14, 22, 45],
    "BH-03": [7, 12, 25, 40, 55],
}
bh_depths = [1.5, 3.0, 4.5, 6.0, 7.5]

# Header
print("Multi-Borehole N-value Summary:")
print(f"  {'BH':>6s}", end="")
for d in bh_depths:
    print(f"  {d:5.1f}m", end="")
print("  | Avg N")
print("  " + "─" * 52)

# Outer loop: each borehole
for bh_name, n_list in boreholes.items():
    avg_n = sum(n_list) / len(n_list)
    print(f"  {bh_name}", end="")
    
    # Inner loop: each layer in this borehole
    for n in n_list:
        print(f"  {n:5d} ", end="")
    
    print(f"  | {avg_n:.1f}")

print()
print("The outer loop runs 3 times (one per borehole).")
print("The inner loop runs 5 times EACH outer loop = 15 total iterations.")
````

## Part 8: Accumulator Patterns — Computing Statistics

Loops are how we compute **statistics** from data — sum, average, min, max, count. These are the building blocks of all data analysis.

| Pattern | Initialize | Update | After Loop |
|---------|-----------|--------|------------|
| Sum | `total = 0` | `total += n` | `total` |
| Count | `count = 0` | `count += 1` | `count` |
| Average | sum + count | — | `total / count` |
| Min | `minimum = first` | `if n < minimum` | `minimum` |
| Max | `maximum = first` | `if n > maximum` | `maximum` |
| Build list | `result = []` | `result.append(x)` | `result` |

````python
# ─────────────────────────────────────────────────────
# ALL ACCUMULATOR PATTERNS IN ONE EXAMPLE
# ─────────────────────────────────────────────────────

n_values = [5, 8, 12, 18, 28, 35, 50, 60]
depths =   [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]

# Initialize accumulators
total_n = 0
n_min = n_values[0]       # start with first value
n_max = n_values[0]
weak_count = 0
vs_list = []              # empty list to fill

# Process each layer
for n in n_values:
    # Sum
    total_n += n
    
    # Min
    if n < n_min:
        n_min = n
    
    # Max
    if n > n_max:
        n_max = n
    
    # Count (conditional)
    if n < 15:
        weak_count += 1
    
    # Build new list (Vs estimation)
    vs = 97 * (n ** 0.314)
    vs_list.append(vs)

# Results
average_n = total_n / len(n_values)

print("BH-01 Statistics:")
print(f"  Sum of N    : {total_n}")
print(f"  Average N   : {average_n:.1f}")
print(f"  Min N       : {n_min}")
print(f"  Max N       : {n_max}")
print(f"  Weak layers : {weak_count} (N < 15)")
print(f"  Vs values   : {[round(v, 1) for v in vs_list]}")

print()
print("Note: In Phase 2 (NumPy), all of this becomes ONE line:")
print("  np.mean(n_values), np.min(), np.max(), etc.")
print("  But understanding the loop version teaches you HOW it works.")
````

## Part 9: Complete Application — Automated Borehole Report

Let's combine **everything** — for loop, enumerate, zip, conditionals, accumulators — into a complete automated borehole report. This is the kind of script that replaces 30 minutes of manual Excel work.

````python
# ══════════════════════════════════════════════════════
# AUTOMATED BOREHOLE REPORT — Combining Everything
# ══════════════════════════════════════════════════════

# Input data (3 parallel lists + site info)
depths =     [1.5, 3.0, 4.5, 6.0, 7.5, 9.0, 10.5, 12.0]
n_values =   [5,   8,   12,  18,  28,  35,  50,   60]
soil_types = ["Fill", "Silty Sand", "Sand", "Clayey Sand",
              "Sand", "Gravel", "Weathered Rock", "Bedrock"]
wt = 3.2

# Accumulators
total_n = 0
liq_warnings = 0
bearing_depth = None       # haven't found it yet
vs_list = []

# Header
print("=" * 80)
print("  AUTOMATED BOREHOLE REPORT — BH-01")
print("  Salt Lake City, Utah | Water Table: 3.2m | Seismic Zone: Yes")
print("=" * 80)
print(f"  {'#':>2s}  {'Depth':>6s}  {'N60':>4s}  {'Soil':16s}  {'Class':13s}  {'Sat':>3s}  {'Vs':>7s}  Liq")
print("  " + "─" * 72)

# Main processing loop: enumerate + zip + conditionals + accumulators
for i, (d, n, s) in enumerate(zip(depths, n_values, soil_types), start=1):
    
    # Consistency (if/elif/else from Day 002)
    if n < 4:
        cons = "Very Loose"
    elif n < 10:
        cons = "Loose"
    elif n < 30:
        cons = "Medium Dense"
    elif n < 50:
        cons = "Dense"
    else:
        cons = "Very Dense"
    
    # Saturation
    sat = "Y" if d > wt else "N"
    
    # Vs estimation (Imai & Tonouchi)
    vs = 97 * (n ** 0.314)
    vs_list.append(vs)
    
    # Liquefaction screening
    if d > wt and n < 20 and "Sand" in s and d < 20:
        liq = "CHECK"
        liq_warnings += 1
    else:
        liq = "OK"
    
    # First bearing layer (break-like logic without break)
    if bearing_depth is None and n >= 15:
        bearing_depth = d
    
    # Accumulate
    total_n += n
    
    # Print row
    print(f"  {i:2d}  {d:5.1f}m  {n:4d}  {s:16s}  {cons:13s}  {sat:>3s}  {vs:6.1f}  {liq}")

# ── Summary Statistics (computed from accumulators) ──
print("  " + "─" * 72)
avg_n = total_n / len(n_values)
avg_vs = sum(vs_list) / len(vs_list)

print(f"\n  STATISTICS:")
print(f"  Average N60 : {avg_n:.1f}")
print(f"  N60 range   : {min(n_values)} — {max(n_values)}")
print(f"  Average Vs  : {avg_vs:.1f} m/s")

print(f"\n  RECOMMENDATIONS:")
if liq_warnings > 0:
    print(f"  [!] {liq_warnings} layer(s) require liquefaction analysis")
else:
    print(f"  [OK] No liquefaction concerns")

if bearing_depth:
    print(f"  [OK] First suitable bearing layer at {bearing_depth:.1f}m")
else:
    print(f"  [!] No suitable bearing layer — consider deep foundation")

print("=" * 80)
````

## Key Takeaways

1. **`for` loops** iterate through collections. Write the logic once, apply to every item. This is the foundation of processing any dataset.

2. **`while` loops** run until a condition is false. Perfect when you don't know how many iterations you need (like drilling until refusal).

3. **`enumerate()`** gives you both index and value — essential for numbered reports and accessing parallel data.

4. **`zip()`** pairs items from multiple lists — this is how you process multi-column data (depth + N + soil together).

5. **Accumulator patterns** (sum, count, min, max, build list) are the manual version of what NumPy/Pandas do automatically. Understanding them helps you understand the libraries.

## How This Connects to ML

| What We Did Today | ML Equivalent |
|-------------------|--------------|
| `for` loop through layers | Iterating through dataset rows |
| `while` until condition | Training loop (`while loss > threshold`) |
| `enumerate` | Batch indexing in DataLoaders |
| `zip(features, labels)` | Pairing X and y in training |
| Accumulator (sum, avg) | Computing loss, accuracy per epoch |
| `break` on condition | Early stopping in training |
| Nested loops | Grid search over hyperparameters |

Every ML training loop looks like this:
```python
for epoch in range(100):           # outer loop (epochs)
    for X_batch, y_batch in data:  # inner loop (batches)
        loss = model(X_batch)      # forward pass
        loss.backward()            # backprop
```
